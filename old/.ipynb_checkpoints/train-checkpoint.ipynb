# plot distribution

```python
import matplotlib.pyplot as plt

# =========================================================
# helper: random subsample for plotting
# =========================================================
def sample_values(x, max_points=300000):
    x = x.detach().cpu().reshape(-1)
    x = x[torch.isfinite(x)]
    if x.numel() > max_points:
        idx = torch.randperm(x.numel())[:max_points]
        x = x[idx]
    return x.numpy()

# =========================================================
# extract land-only target values
# =========================================================
y_train_land = y_train[mask_train > 0.5]
y_val_land   = y_val[mask_val > 0.5]

# normalized target over land
y_train_n_land = y_train_n[mask_train > 0.5]
y_val_n_land   = y_val_n[mask_val > 0.5]

# optional: also look at coarse input distribution
X_train_sample = sample_values(X_train)
X_val_sample   = sample_values(X_val)

y_train_sample = sample_values(y_train_land)
y_val_sample   = sample_values(y_val_land)

y_train_n_sample = sample_values(y_train_n_land)
y_val_n_sample   = sample_values(y_val_n_land)

# for precipitation, log1p helps visualize the tail
if rcm_var == "prec":
    y_train_log = np.log1p(np.clip(y_train_sample, a_min=0, a_max=None))
    y_val_log   = np.log1p(np.clip(y_val_sample,   a_min=0, a_max=None))

# =========================================================
# plot
# =========================================================
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# ---- coarse input X ----
bins_x = 100
xmin = min(X_train_sample.min(), X_val_sample.min())
xmax = max(X_train_sample.max(), X_val_sample.max())

axes[0, 0].hist(X_train_sample, bins=bins_x, range=(xmin, xmax),
                density=True, alpha=0.5, label="Train")
axes[0, 0].hist(X_val_sample, bins=bins_x, range=(xmin, xmax),
                density=True, alpha=0.5, label="Val")
axes[0, 0].set_title("Input X distribution")
axes[0, 0].set_xlabel("X")
axes[0, 0].set_ylabel("Density")
axes[0, 0].legend()

# ---- raw target y over land ----
# use percentile cap for prec so the histogram is not dominated by extremes
if rcm_var == "prec":
    upper = np.percentile(np.concatenate([y_train_sample, y_val_sample]), 99.5)
    yrange = (0, upper)
else:
    ymin = min(y_train_sample.min(), y_val_sample.min())
    ymax = max(y_train_sample.max(), y_val_sample.max())
    yrange = (ymin, ymax)

axes[0, 1].hist(y_train_sample, bins=100, range=yrange,
                density=True, alpha=0.5, label="Train")
axes[0, 1].hist(y_val_sample, bins=100, range=yrange,
                density=True, alpha=0.5, label="Val")
axes[0, 1].set_title("Target y distribution over land")
axes[0, 1].set_xlabel("y")
axes[0, 1].set_ylabel("Density")
axes[0, 1].legend()

# ---- normalized target y over land ----
yn_min = min(y_train_n_sample.min(), y_val_n_sample.min())
yn_max = max(y_train_n_sample.max(), y_val_n_sample.max())

axes[1, 0].hist(y_train_n_sample, bins=100, range=(yn_min, yn_max),
                density=True, alpha=0.5, label="Train")
axes[1, 0].hist(y_val_n_sample, bins=100, range=(yn_min, yn_max),
                density=True, alpha=0.5, label="Val")
axes[1, 0].set_title("Normalized target y distribution over land")
axes[1, 0].set_xlabel("normalized y")
axes[1, 0].set_ylabel("Density")
axes[1, 0].legend()

# ---- log1p precipitation ----
if rcm_var == "prec":
    log_min = min(y_train_log.min(), y_val_log.min())
    log_max = max(y_train_log.max(), y_val_log.max())

    axes[1, 1].hist(y_train_log, bins=100, range=(log_min, log_max),
                    density=True, alpha=0.5, label="Train")
    axes[1, 1].hist(y_val_log, bins=100, range=(log_min, log_max),
                    density=True, alpha=0.5, label="Val")
    axes[1, 1].set_title("log1p(target y) distribution over land")
    axes[1, 1].set_xlabel("log1p(y)")
    axes[1, 1].set_ylabel("Density")
    axes[1, 1].legend()
else:
    axes[1, 1].axis("off")

plt.tight_layout()
plt.show()
```
