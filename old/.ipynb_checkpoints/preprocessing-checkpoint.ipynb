# RCM only (interpolation)

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np
import xarray as xr
import pandas as pd
import torch
import torch.nn.functional as F

# check current env
import os
import glob
```

```python
rcm_var = "tmean"
gcm_name = "CanESM2"
rcm_name = "CanRCM4"
grid = 'NAM-44i'
factor = 4   # 4
gcm_rcm_dict = {gcm_name: rcm_name}

rcm_path = f"/projects/sds-lab/Shuochen/downscaling/CORDEX/{rcm_var}.*.{gcm_name}.{gcm_rcm_dict[gcm_name]}.day.{grid}.mbcn-gridMET.nc"

rcm = xr.open_mfdataset(rcm_path, combine="by_coords")[rcm_var]
# select CONUS
lat_min, lat_max = 24, 50
lon_min, lon_max = -125, -66
rcm = rcm.sel(lat=slice(lat_min,lat_max), lon=slice(lon_min,lon_max))



# =========================================================
# create output folder
# =========================================================
out_dir = f"/projects/sds-lab/Shuochen/downscaling/CORDEX/{rcm_var}.{gcm_name}.{gcm_rcm_dict[gcm_name]}.day.{grid}.mbcn-gridMET"

os.makedirs(out_dir, exist_ok=True)
# =========================================================
# input
# =========================================================
da = rcm.copy()

out_high = f"{out_dir}/high_res.pth"
out_low  = f"{out_dir}/coarse_{factor}x.pth"
out_mask_high = f"{out_dir}/high_res_mask.pth"
out_mask_low  = f"{out_dir}/coarse_{factor}x_mask.pth"
# =========================================================
# find lat/lon names
# =========================================================
lat_name, lon_name = None, None

for name in da.coords:
    low = name.lower()
    if low in ["lat", "latitude"]:
        lat_name = name
    elif low in ["lon", "longitude"]:
        lon_name = name

if lat_name is None or lon_name is None:
    raise ValueError("Could not find lat/lon coordinates.")

if lat_name not in da.dims or lon_name not in da.dims:
    raise ValueError("lat/lon must be dimensions.")

lat = np.asarray(da[lat_name].values)
lon = np.asarray(da[lon_name].values)

# =========================================================
# move to (..., lat, lon)
# =========================================================
other_dims = [d for d in da.dims if d not in [lat_name, lon_name]]
da = da.transpose(*other_dims, lat_name, lon_name)

x = da.values

# =========================================================
# 🔥 CREATE LAND MASK (before filling NaNs)
# =========================================================
mask = (~np.isnan(x)).astype(np.float32)   # 1=land, 0=sea

# =========================================================
# fill NaNs with zero
# =========================================================
x = np.nan_to_num(x, nan=0.0)

H, W = x.shape[-2], x.shape[-1]

# =========================================================
# enforce divisibility by factor
# =========================================================
Hf = (H // factor) * factor
Wf = (W // factor) * factor

if Hf == 0 or Wf == 0:
    raise ValueError("Grid too small after enforcing factor.")

trim_lat = H - Hf
trim_lon = W - Wf

top = trim_lat // 2
bottom = top + Hf
left = trim_lon // 2
right = left + Wf

# apply SAME trim to mask
x = x[..., top:bottom, left:right]
mask = mask[..., top:bottom, left:right]

lat_trim = lat[top:bottom]
lon_trim = lon[left:right]

print(f"Original shape: ({H}, {W})")
print(f"Trimmed high-res shape: ({Hf}, {Wf})")
print(f"Coarse shape: ({Hf//factor}, {Wf//factor})")
print(f"Factor: {factor}")

# =========================================================
# reshape for PyTorch
# =========================================================
lead_shape = x.shape[:-2]
N = int(np.prod(lead_shape)) if len(lead_shape) > 0 else 1

x_4d    = x.reshape(N, 1, Hf, Wf)
mask_4d = mask.reshape(N, 1, Hf, Wf)

x_torch    = torch.as_tensor(x_4d, dtype=torch.float32)
mask_torch = torch.as_tensor(mask_4d, dtype=torch.float32)

# =========================================================
# bilinear downsample (data)
# =========================================================
y_torch = F.interpolate(
    x_torch,
    size=(Hf // factor, Wf // factor),
    mode="bilinear",
    align_corners=False
)

# =========================================================
# 🔥 nearest downsample (mask — VERY IMPORTANT)
# =========================================================
mask_low_torch = F.interpolate(
    mask_torch,
    size=(Hf // factor, Wf // factor),
    mode="nearest"
)

# reshape back
x_high   = x_torch.reshape(*lead_shape, 1, Hf, Wf).contiguous()
x_low    = y_torch.reshape(*lead_shape, 1, Hf // factor, Wf // factor).contiguous()

mask_high = mask_torch.reshape(*lead_shape, 1, Hf, Wf).contiguous()
mask_low  = mask_low_torch.reshape(*lead_shape, 1, Hf // factor, Wf // factor).contiguous()

# =========================================================
# save tensors
# =========================================================
torch.save(x_high, out_high)
torch.save(x_low,  out_low)

torch.save(mask_high, out_mask_high)
torch.save(mask_low,  out_mask_low)

print(f"Saved high-res data: {out_high}")
print(f"Saved coarse data:   {out_low}")
print(f"Saved high-res mask: {out_mask_high}")
print(f"Saved coarse mask:   {out_mask_low}")

# =========================================================
# sanity check
# =========================================================
assert x_high.shape[-2] == factor * x_low.shape[-2]
assert x_high.shape[-1] == factor * x_low.shape[-1]

print("✔ Shape consistency check passed")

# =========================================================
# optional info
# =========================================================
print("Lat range:", lat_trim[0], "to", lat_trim[-1], "n =", len(lat_trim))
print("Lon range:", lon_trim[0], "to", lon_trim[-1], "n =", len(lon_trim))

print("Land fraction (high-res):", mask_high.mean().item())
print("Land fraction (coarse): ", mask_low.mean().item())
```

# GCM and RCM (new)

```python
import os
import glob
import json
import numpy as np
import torch
import torch.nn.functional as F
import xarray as xr

import rasterio
from rasterio.warp import reproject, Resampling
from rasterio.transform import from_bounds


# =========================================================
# settings
# =========================================================
rcm_var = "tmean"      # HR RCM target variable
gcm_var = "tas"        # GCM input variable

gcm_name = "CanESM2"
rcm_name = "CanRCM4"
grid = "NAM-44i"

factor = 4

lat_min, lat_max = 24, 50
lon_min, lon_max = -125, -66

gcm_dir = "/projects/sds-lab/Shuochen/downscaling/GCM"

# GMTED2010 mean-elevation ArcGrid folder
elev_raster_path = "/projects/sds-lab/Shuochen/downscaling/mn30_grd"

rcm_path = (
    f"/projects/sds-lab/Shuochen/downscaling/CORDEX/"
    f"{rcm_var}.*.{gcm_name}.{rcm_name}.day.{grid}.mbcn-gridMET.nc"
)

gcm_path = (
    f"{gcm_dir}/"
    f"{gcm_var}_day_{gcm_name}_*_r1i1p1_*.nc"
)

out_dir = (
    f"/projects/sds-lab/Shuochen/downscaling/CORDEX/"
    f"{rcm_var}.{gcm_name}.{rcm_name}.day.{grid}.GCM_to_HR_RCM"
)

os.makedirs(out_dir, exist_ok=True)


# =========================================================
# output files
# =========================================================
out_high = f"{out_dir}/high_res.pth"                 # HR RCM
out_low = f"{out_dir}/low_res.pth"                   # GCM on LR grid
out_coarse = f"{out_dir}/coarse_{factor}x.pth"       # coarsened RCM

out_mask_high = f"{out_dir}/high_res_mask.pth"
out_mask_low = f"{out_dir}/low_res_mask.pth"
out_mask_coarse = f"{out_dir}/coarse_{factor}x_mask.pth"

# elevation outputs
out_elev_high = f"{out_dir}/high_res_elevation.pth"
out_elev_low = f"{out_dir}/low_res_elevation.pth"
out_elev_coarse = f"{out_dir}/coarse_{factor}x_elevation.pth"
out_elev_stats = f"{out_dir}/elevation_stats.json"


# =========================================================
# helper functions
# =========================================================
def find_lat_lon_names(da):
    lat_name, lon_name = None, None

    for name in da.coords:
        low = name.lower()
        if low in ["lat", "latitude"]:
            lat_name = name
        elif low in ["lon", "longitude"]:
            lon_name = name

    if lat_name is None or lon_name is None:
        raise ValueError("Could not find lat/lon coordinates.")

    if lat_name not in da.dims or lon_name not in da.dims:
        raise ValueError("lat/lon must be dimensions.")

    return lat_name, lon_name


def convert_lon_to_minus180_180(da, lon_name):
    lon = da[lon_name]

    if lon.max() > 180:
        da = da.assign_coords(
            {lon_name: (((lon + 180) % 360) - 180)}
        )
        da = da.sortby(lon_name)

    return da


def standardize_time_to_dates(da):
    """
    Convert time coordinate to YYYY-MM-DD strings.
    This helps align RCM and GCM even if calendars differ slightly.
    """
    if "time" not in da.dims:
        raise ValueError("Expected time dimension in data.")

    dates = da["time"].dt.strftime("%Y-%m-%d")

    da = da.assign_coords(date=("time", dates.data))
    da = da.swap_dims({"time": "date"})
    da = da.drop_vars("time")
    da = da.rename({"date": "time"})

    return da


def maybe_convert_gcm_units(gcm, rcm):
    """
    Convert GCM tas units to match RCM tmean units when units metadata is available.
    CanESM2 tas is often in K. Some downscaling targets may be in degC.
    """
    gcm_units = str(gcm.attrs.get("units", "")).lower()
    rcm_units = str(rcm.attrs.get("units", "")).lower()

    gcm_is_k = ("k" == gcm_units) or ("kelvin" in gcm_units)
    rcm_is_k = ("k" == rcm_units) or ("kelvin" in rcm_units)

    gcm_is_c = ("c" == gcm_units) or ("celsius" in gcm_units) or ("degc" in gcm_units)
    rcm_is_c = ("c" == rcm_units) or ("celsius" in rcm_units) or ("degc" in rcm_units)

    if gcm_is_k and rcm_is_c:
        print("Converting GCM tas from K to degC.")
        gcm = gcm - 273.15
        gcm.attrs["units"] = rcm.attrs.get("units", "degC")

    elif gcm_is_c and rcm_is_k:
        print("Converting GCM tas from degC to K.")
        gcm = gcm + 273.15
        gcm.attrs["units"] = rcm.attrs.get("units", "K")

    else:
        print("No automatic unit conversion applied.")
        print("GCM units:", gcm_units)
        print("RCM units:", rcm_units)

    return gcm


def create_elevation_from_arcgrid(
    raster_path,
    lat_1d,
    lon_1d,
    land_mask_2d=None,
    resampling="average",
):
    """
    Read GMTED2010 ESRI ArcGrid and resample it onto the target lat/lon grid.

    raster_path should be the ArcGrid folder, e.g.
    /projects/sds-lab/Shuochen/downscaling/mn30_grd

    Returns
    -------
    elev : np.ndarray
        Elevation array with shape [lat, lon].
    """
    H = len(lat_1d)
    W = len(lon_1d)

    if H < 2 or W < 2:
        raise ValueError("lat_1d and lon_1d must have at least 2 points.")

    dx = float(abs(lon_1d[1] - lon_1d[0]))
    dy = float(abs(lat_1d[1] - lat_1d[0]))

    lon_min_grid = float(np.min(lon_1d) - dx / 2)
    lon_max_grid = float(np.max(lon_1d) + dx / 2)
    lat_min_grid = float(np.min(lat_1d) - dy / 2)
    lat_max_grid = float(np.max(lat_1d) + dy / 2)

    dst_transform = from_bounds(
        lon_min_grid,
        lat_min_grid,
        lon_max_grid,
        lat_max_grid,
        W,
        H,
    )

    if resampling == "average":
        resampling_method = Resampling.average
    elif resampling == "bilinear":
        resampling_method = Resampling.bilinear
    elif resampling == "nearest":
        resampling_method = Resampling.nearest
    else:
        raise ValueError("resampling must be 'average', 'bilinear', or 'nearest'.")

    with rasterio.open(raster_path) as src:
        print("\nElevation raster opened:")
        print("Path:", raster_path)
        print("CRS:", src.crs)
        print("Bounds:", src.bounds)
        print("Shape:", src.height, src.width)
        print("Nodata:", src.nodata)

        src_crs = src.crs
        if src_crs is None:
            print("Warning: elevation raster CRS is missing. Assuming EPSG:4326.")
            src_crs = "EPSG:4326"

        elev_north_to_south = np.full((H, W), np.nan, dtype=np.float32)

        reproject(
            source=rasterio.band(src, 1),
            destination=elev_north_to_south,
            src_transform=src.transform,
            src_crs=src_crs,
            src_nodata=src.nodata,
            dst_transform=dst_transform,
            dst_crs="EPSG:4326",
            dst_nodata=np.nan,
            resampling=resampling_method,
        )

    # rasterio output is north-to-south.
    # RCM grid is sorted south-to-north after sortby(lat).
    if lat_1d[0] < lat_1d[-1]:
        elev = np.flipud(elev_north_to_south)
    else:
        elev = elev_north_to_south

    elev = elev.astype(np.float32)
    elev = np.nan_to_num(elev, nan=0.0)

    if land_mask_2d is not None:
        elev = elev * land_mask_2d.astype(np.float32)

    return elev


# =========================================================
# check input files
# =========================================================
rcm_files = sorted(glob.glob(rcm_path))
gcm_files = sorted(glob.glob(gcm_path))

if len(rcm_files) == 0:
    raise FileNotFoundError(f"No RCM files found with pattern:\n{rcm_path}")

if len(gcm_files) == 0:
    raise FileNotFoundError(f"No GCM files found with pattern:\n{gcm_path}")

print("Number of RCM files:", len(rcm_files))
print("Number of GCM files:", len(gcm_files))


# =========================================================
# open HR RCM target
# =========================================================
# rcm = xr.open_mfdataset(
#     rcm_files,
#     combine="nested",
#     concat_dim="time",
#     parallel=False,
#     data_vars="minimal",
#     coords="minimal",
#     compat="override",
# )[rcm_var]
rcm = xr.open_mfdataset(rcm_files)[rcm_var]

rcm_lat_name, rcm_lon_name = find_lat_lon_names(rcm)

rcm = convert_lon_to_minus180_180(rcm, rcm_lon_name)

rcm = rcm.sel(
    {
        rcm_lat_name: slice(lat_min, lat_max),
        rcm_lon_name: slice(lon_min, lon_max),
    }
)

rcm = rcm.sortby(rcm_lat_name).sortby(rcm_lon_name)


# =========================================================
# open GCM input
# =========================================================
# gcm = xr.open_mfdataset(
#     gcm_files,
#     combine="nested",
#     concat_dim="time",
#     parallel=False,
#     data_vars="minimal",
#     coords="minimal",
#     compat="override",
# )[gcm_var]
gcm = xr.open_mfdataset(gcm_files)[gcm_var]

gcm_lat_name, gcm_lon_name = find_lat_lon_names(gcm)

gcm = convert_lon_to_minus180_180(gcm, gcm_lon_name)
gcm = gcm.sortby(gcm_lat_name).sortby(gcm_lon_name)

# select slightly larger GCM region before interpolation
buffer_deg = 5

gcm = gcm.sel(
    {
        gcm_lat_name: slice(lat_min - buffer_deg, lat_max + buffer_deg),
        gcm_lon_name: slice(lon_min - buffer_deg, lon_max + buffer_deg),
    }
)

gcm = maybe_convert_gcm_units(gcm, rcm)


# =========================================================
# trim HR RCM so H and W are divisible by factor
# =========================================================
rcm_lat = np.asarray(rcm[rcm_lat_name].values)
rcm_lon = np.asarray(rcm[rcm_lon_name].values)

other_dims = [d for d in rcm.dims if d not in [rcm_lat_name, rcm_lon_name]]
rcm = rcm.transpose(*other_dims, rcm_lat_name, rcm_lon_name)

H, W = rcm.shape[-2], rcm.shape[-1]

Hf = (H // factor) * factor
Wf = (W // factor) * factor

if Hf == 0 or Wf == 0:
    raise ValueError("Grid too small after enforcing factor.")

trim_lat = H - Hf
trim_lon = W - Wf

top = trim_lat // 2
bottom = top + Hf

left = trim_lon // 2
right = left + Wf

rcm = rcm.isel(
    {
        rcm_lat_name: slice(top, bottom),
        rcm_lon_name: slice(left, right),
    }
)

lat_trim = np.asarray(rcm[rcm_lat_name].values)
lon_trim = np.asarray(rcm[rcm_lon_name].values)

print(f"Original HR RCM shape: ({H}, {W})")
print(f"Trimmed HR RCM shape: ({Hf}, {Wf})")
print(f"LR shape: ({Hf // factor}, {Wf // factor})")
print(f"Factor: {factor}")


# =========================================================
# define LR grid from HR RCM grid
# =========================================================
# This grid is used for both:
# 1. GCM low_res
# 2. coarsened RCM coarse_4x
lr_lat = lat_trim[factor // 2 :: factor]
lr_lon = lon_trim[factor // 2 :: factor]

lr_lat = lr_lat[: Hf // factor]
lr_lon = lr_lon[: Wf // factor]

print("LR lat size:", len(lr_lat))
print("LR lon size:", len(lr_lon))


# =========================================================
# regrid GCM tas to LR grid
# =========================================================
gcm_lr = gcm.interp(
    {
        gcm_lat_name: xr.DataArray(lr_lat, dims="lat_lr"),
        gcm_lon_name: xr.DataArray(lr_lon, dims="lon_lr"),
    },
    method="linear",
)

# Keep LR spatial dimensions different from HR RCM dimensions.
# Do NOT rename lat_lr/lon_lr to lat/lon before time alignment.
gcm_lr = gcm_lr.transpose("time", "lat_lr", "lon_lr")


# =========================================================
# align time only between GCM input and RCM target
# =========================================================
rcm_time_aligned = standardize_time_to_dates(rcm)
gcm_time_aligned = standardize_time_to_dates(gcm_lr)

# Drop possible duplicate dates
rcm_time_aligned = rcm_time_aligned.isel(
    time=~rcm_time_aligned.get_index("time").duplicated()
)

gcm_time_aligned = gcm_time_aligned.isel(
    time=~gcm_time_aligned.get_index("time").duplicated()
)

# Align by common dates only
common_times = np.intersect1d(
    rcm_time_aligned["time"].values,
    gcm_time_aligned["time"].values,
)

if len(common_times) == 0:
    raise ValueError("No overlapping dates found between RCM and GCM.")

rcm_time_aligned = rcm_time_aligned.sel(time=common_times)
gcm_time_aligned = gcm_time_aligned.sel(time=common_times)

print("Aligned samples:", rcm_time_aligned.sizes["time"])
print("First aligned date:", rcm_time_aligned["time"].values[0])
print("Last aligned date:", rcm_time_aligned["time"].values[-1])

print("RCM aligned shape:", rcm_time_aligned.shape)
print("GCM aligned shape:", gcm_time_aligned.shape)


# =========================================================
# load arrays
# =========================================================
# Cast to float32 before loading to reduce memory.
print("Casting aligned arrays to float32...")
rcm_time_aligned = rcm_time_aligned.astype("float32")
gcm_time_aligned = gcm_time_aligned.astype("float32")

print("Loading HR RCM array...")
hr = rcm_time_aligned.load().values          # [time, H, W]

print("Loading GCM low_res array...")
low = gcm_time_aligned.load().values         # [time, h, w]


# =========================================================
# masks before filling NaNs
# =========================================================
# Full time dimension masks
mask_high = np.isfinite(hr).astype(np.float32)
mask_low = np.isfinite(low).astype(np.float32)

# Fill NaNs with zero
hr = np.nan_to_num(hr, nan=0.0, copy=False)
low = np.nan_to_num(low, nan=0.0, copy=False)

print("HR RCM array shape:", hr.shape)
print("GCM low_res array shape:", low.shape)
print("HR mask shape:", mask_high.shape)
print("GCM low_res mask shape:", mask_low.shape)


# =========================================================
# reshape HR RCM and GCM low_res to PyTorch format
# =========================================================
lead_shape_hr = hr.shape[:-2]
lead_shape_low = low.shape[:-2]

N_hr = int(np.prod(lead_shape_hr))
N_low = int(np.prod(lead_shape_low))

if N_hr != N_low:
    raise ValueError(f"HR and GCM sample counts differ: {N_hr} vs {N_low}")

hr_4d = hr.reshape(N_hr, 1, Hf, Wf)
low_4d = low.reshape(N_low, 1, Hf // factor, Wf // factor)

mask_high_4d = mask_high.reshape(N_hr, 1, Hf, Wf)
mask_low_4d = mask_low.reshape(N_low, 1, Hf // factor, Wf // factor)

x_high = torch.as_tensor(hr_4d, dtype=torch.float32).contiguous()
x_low = torch.as_tensor(low_4d, dtype=torch.float32).contiguous()

mask_high_torch = torch.as_tensor(mask_high_4d, dtype=torch.float32).contiguous()
mask_low_torch = torch.as_tensor(mask_low_4d, dtype=torch.float32).contiguous()


# =========================================================
# create coarsened RCM using bilinear downsampling
# =========================================================
# This is the old RCM-to-RCM low-resolution input.
# It is saved separately as coarse_4x.pth.
x_coarse = F.interpolate(
    x_high,
    size=(Hf // factor, Wf // factor),
    mode="bilinear",
    align_corners=False,
)

mask_coarse = F.interpolate(
    mask_high_torch,
    size=(Hf // factor, Wf // factor),
    mode="nearest",
)

x_coarse = x_coarse.contiguous()
mask_coarse = mask_coarse.contiguous()

print("Coarsened RCM shape:", x_coarse.shape)
print("Coarsened RCM mask shape:", mask_coarse.shape)


# =========================================================
# create elevation tensors
# =========================================================
# Use first HR mask for spatial land mask when creating terrain.
static_hr_land_mask = mask_high_4d[0, 0, :, :]  # [Hf, Wf]

elev_high_2d = create_elevation_from_arcgrid(
    raster_path=elev_raster_path,
    lat_1d=lat_trim,
    lon_1d=lon_trim,
    land_mask_2d=static_hr_land_mask,
    resampling="average",
)

elev_high = torch.as_tensor(
    elev_high_2d[None, None, :, :],
    dtype=torch.float32,
).contiguous()

static_hr_land_mask_torch = torch.as_tensor(
    static_hr_land_mask[None, None, :, :],
    dtype=torch.float32,
)

# Normalize elevation over HR land pixels only
land_elev = elev_high[static_hr_land_mask_torch > 0.5]

if land_elev.numel() == 0:
    raise ValueError("No valid land elevation pixels found.")

elev_mean = land_elev.mean()
elev_std = land_elev.std()

if elev_std.item() == 0:
    elev_std = torch.tensor(1.0, dtype=torch.float32)

elev_high_n_static = (elev_high - elev_mean) / elev_std
elev_high_n_static = elev_high_n_static * static_hr_land_mask_torch

# Downsample HR elevation to the shared LR grid.
# This LR grid is used by both low_res.pth and coarse_4x.pth.
elev_low_n_static = F.interpolate(
    elev_high_n_static,
    size=(Hf // factor, Wf // factor),
    mode="bilinear",
    align_corners=False,
).contiguous()

# Zero ocean in LR elevation using the first coarsened HR land mask
static_lr_land_mask = mask_coarse[0:1]  # [1, 1, h, w]
elev_low_n_static = elev_low_n_static * static_lr_land_mask

# low_res and coarse_4x have the same LR grid, so elevation is identical
elev_coarse_n_static = elev_low_n_static.clone().contiguous()

# =========================================================
# duplicate elevation along full time/sample dimension
# =========================================================
elev_high_n = elev_high_n_static.expand(N_hr, -1, -1, -1).contiguous()
elev_low_n = elev_low_n_static.expand(N_low, -1, -1, -1).contiguous()
elev_coarse_n = elev_coarse_n_static.expand(N_hr, -1, -1, -1).contiguous()

print("Elevation HR static shape:", elev_high_n_static.shape)
print("Elevation LR static shape:", elev_low_n_static.shape)
print("Elevation HR full shape:", elev_high_n.shape)
print("Elevation LR full shape:", elev_low_n.shape)
print("Elevation mean/std over HR land:", elev_mean.item(), elev_std.item())


# =========================================================
# save tensors
# =========================================================
torch.save(x_high, out_high)
torch.save(x_low, out_low)
torch.save(x_coarse, out_coarse)

torch.save(mask_high_torch, out_mask_high)
torch.save(mask_low_torch, out_mask_low)
torch.save(mask_coarse, out_mask_coarse)

torch.save(elev_high_n, out_elev_high)
torch.save(elev_low_n, out_elev_low)
torch.save(elev_coarse_n, out_elev_coarse)

elev_stats = {
    "elevation_source": elev_raster_path,
    "elevation_product": "GMTED2010 mn30_grd mean elevation, 30 arc-second",
    "normalization": "z-score over HR RCM land pixels",
    "elev_mean": float(elev_mean.item()),
    "elev_std": float(elev_std.item()),
    "high_res_elevation_shape": list(elev_high_n.shape),
    "low_res_elevation_shape": list(elev_low_n.shape),
    "coarse_elevation_shape": list(elev_coarse_n.shape),
    "high_res_mask_shape": list(mask_high_torch.shape),
    "low_res_mask_shape": list(mask_low_torch.shape),
    "coarse_mask_shape": list(mask_coarse.shape),
}

with open(out_elev_stats, "w") as f:
    json.dump(elev_stats, f, indent=4)

print(f"Saved HR RCM target:          {out_high}")
print(f"Saved GCM low_res input:      {out_low}")
print(f"Saved coarsened RCM input:    {out_coarse}")

print(f"Saved HR RCM mask:            {out_mask_high}")
print(f"Saved GCM low_res mask:       {out_mask_low}")
print(f"Saved coarsened RCM mask:     {out_mask_coarse}")

print(f"Saved HR elevation:           {out_elev_high}")
print(f"Saved GCM low_res elevation:  {out_elev_low}")
print(f"Saved coarse RCM elevation:   {out_elev_coarse}")
print(f"Saved elevation stats:        {out_elev_stats}")


# =========================================================
# sanity checks
# =========================================================
assert x_high.shape[0] == x_low.shape[0]
assert x_high.shape[0] == x_coarse.shape[0]

assert x_high.shape[-2] == factor * x_low.shape[-2]
assert x_high.shape[-1] == factor * x_low.shape[-1]

assert x_high.shape[-2] == factor * x_coarse.shape[-2]
assert x_high.shape[-1] == factor * x_coarse.shape[-1]

assert x_low.shape == x_coarse.shape

# Full-time mask checks
assert mask_high_torch.shape == x_high.shape
assert mask_low_torch.shape == x_low.shape
assert mask_coarse.shape == x_coarse.shape

# Full-time elevation checks
assert elev_high_n.shape == x_high.shape
assert elev_low_n.shape == x_low.shape
assert elev_coarse_n.shape == x_coarse.shape

print("✔ Shape consistency check passed")

print("HR tensor shape:", x_high.shape)
print("GCM low_res tensor shape:", x_low.shape)
print("Coarsened RCM tensor shape:", x_coarse.shape)

print("HR mask tensor shape:", mask_high_torch.shape)
print("GCM low_res mask tensor shape:", mask_low_torch.shape)
print("Coarsened RCM mask tensor shape:", mask_coarse.shape)

print("HR elevation tensor shape:", elev_high_n.shape)
print("LR elevation tensor shape:", elev_low_n.shape)
print("Coarse elevation tensor shape:", elev_coarse_n.shape)

print("Lat range HR:", lat_trim[0], "to", lat_trim[-1], "n =", len(lat_trim))
print("Lon range HR:", lon_trim[0], "to", lon_trim[-1], "n =", len(lon_trim))

print("Lat range LR:", lr_lat[0], "to", lr_lat[-1], "n =", len(lr_lat))
print("Lon range LR:", lr_lon[0], "to", lr_lon[-1], "n =", len(lr_lon))

print("Valid fraction high_res RCM:", mask_high_torch.mean().item())
print("Valid fraction GCM low_res:", mask_low_torch.mean().item())
print("Valid fraction coarse RCM:", mask_coarse.mean().item())

print("Elevation normalized min/max HR:", elev_high_n.min().item(), elev_high_n.max().item())
print("Elevation normalized min/max LR:", elev_low_n.min().item(), elev_low_n.max().item())
```

# plot

```python
# =========================================================
# load tensors from folder
# =========================================================
high_res = 0.25   # change to 0.25 if needed

out_dir = f"/projects/sds-lab/Shuochen/downscaling/CORDEX/{rcm_var}.{gcm_name}.{gcm_rcm_dict[gcm_name]}.day.{grid}.mbcn-gridMET"

x_high    = torch.load(f"{out_dir}/high_res.pth")
x_low     = torch.load(f"{out_dir}/coarse_{factor}x.pth")
mask_high = torch.load(f"{out_dir}/high_res_mask.pth")
mask_low  = torch.load(f"{out_dir}/coarse_{factor}x_mask.pth")

x_high = x_high.float()
x_low = x_low.float()
mask_high = mask_high.float()
mask_low = mask_low.float()

print(f"High-res ({high_res}°) shape:", x_high.shape)
print(f"Coarse (factor={factor}) shape:", x_low.shape)
print("High-res mask shape:", mask_high.shape)
print("Coarse mask shape:", mask_low.shape)

# =========================================================
# resolution label
# =========================================================
target_res = high_res * factor

# =========================================================
# pick a sample
# =========================================================
def get_first_2d(x):
    while x.ndim > 2:
        x = x[0]
    return x

f_high = get_first_2d(x_high).numpy()
f_low = get_first_2d(x_low).numpy()
m_high = get_first_2d(mask_high).numpy()
m_low = get_first_2d(mask_low).numpy()

# =========================================================
# upsample coarse field and coarse mask → high-res
# =========================================================
f_low_up = F.interpolate(
    torch.tensor(f_low, dtype=torch.float32)[None, None],
    size=f_high.shape,
    mode="bilinear",
    align_corners=False
)[0, 0].numpy()

m_low_up = F.interpolate(
    torch.tensor(m_low, dtype=torch.float32)[None, None],
    size=f_high.shape,
    mode="nearest"
)[0, 0].numpy()

# =========================================================
# apply mask for plotting
# =========================================================
f_high_plot = np.where(m_high > 0.5, f_high, np.nan)
f_low_up_plot = np.where(m_high > 0.5, f_low_up, np.nan)
diff = np.where(m_high > 0.5, f_high - f_low_up, np.nan)

# =========================================================
# plot
# =========================================================
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

vmin = np.nanmin(f_high_plot)
vmax = np.nanmax(f_high_plot)

im0 = axes[0].imshow(f_high_plot, origin="lower", vmin=vmin, vmax=vmax)
axes[0].set_title(f"{high_res}° (original, land only)")

im1 = axes[1].imshow(f_low_up_plot, origin="lower", vmin=vmin, vmax=vmax)
axes[1].set_title(f"{target_res:.2f}° → {high_res}° (upsampled, land only)")

diff_lim = np.nanmax(np.abs(diff))
im2 = axes[2].imshow(diff, origin="lower", cmap="RdBu_r", vmin=-diff_lim, vmax=diff_lim)
axes[2].set_title(f"Difference ({high_res}° - upsampled {target_res:.2f}°)")

plt.colorbar(im0, ax=axes[0], fraction=0.046)
plt.colorbar(im1, ax=axes[1], fraction=0.046)
plt.colorbar(im2, ax=axes[2], fraction=0.046)

plt.tight_layout()
plt.show()
```
