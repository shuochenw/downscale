```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np
import xarray as xr
import xesmf as xe
import pandas as pd

# check current env
import os
print(os.environ['CONDA_DEFAULT_ENV'])
import glob
```

# regrid all files in folder

```python
# specify folder and target
folders = '/projects/sds-lab/Shuochen/downscaling/canesm2/'
res = 2
targets = xr.Dataset(
    {
        "lat": (["lat"], np.arange(-90, 90 + res, res), {"units": "degrees_north"}),
        "lon": (["lon"], np.arange(-180, 180 + res,  + res), {"units": "degrees_east"}),
    }
)
# targets = xr.open_dataset('/work/sds-lab/Shuochen/CMIP/sst.mon.mean.nc')
methods = 'bilinear'


def regrid(folder, target, method):
    """Convenience function for one-time regridding"""
    # list files in folder
    file_full_path = glob.glob(folder + '*.nc')
    # print(file_full_path)
    allfiles = [ f for f in os.listdir(folder) if os.path.isfile(os.path.join(folder,f)) ]
    print(f'Total files: {len(allfiles)}')
    print('-----------------------')
    # make a new sub-dir called "regrid"
    directory = "regrid/"
    path = os.path.join(folder, directory)
    if os.path.exists(path) == False:
        os.mkdir(path)
    # define and check info of the regridder (of the first file)
    regridder_info = xe.Regridder(xr.open_dataset(file_full_path[0]), target, method)
    print(regridder_info)
    print('-----------------------')
    
    for i in range(len(file_full_path)):
        if os.path.exists(path + 'regridded_' + allfiles[i]) == False:
            regridder = xe.Regridder(xr.open_dataset(file_full_path[i]), target, method)
            ds_out = regridder(xr.open_dataset(file_full_path[i]))
            ds_out.to_netcdf(path + 'regridded_' + allfiles[i])
            print(path + 'regridded_' + allfiles[i])
        
    return 'Complete'

# call function
regrid(folders,targets,methods)
```

# GCM (to 2) -> RCM (to 0.5)

```python
var = 'tas'
gcm_name = 'CanESM2'
rcm_name = 'RCA4'
lat_min = 24
lat_max = 50
lon_min = -124
lon_max = -66
gcm_res = 2
rcm_res = 0.5
gcm_start = '1850-01-01'
gcm_end = '2100-12-31'
rcm_start = '1951-01-01'
rcm_end = '2100-12-31'
prism_elevation = '/projects/sds-lab/Shuochen/downscaling/elevation.nc'
scaling_factor = 4
if scaling_factor == 4:
    offset = 0.5 + 0.25

gcm = xr.open_mfdataset(f'/projects/sds-lab/Shuochen/downscaling/canesm2/{var}_day_{gcm_name}_*.nc')
# fix lon coord
gcm = gcm.assign_coords(lon=(((gcm.lon + 180) % 360) - 180)).sortby('lon')
# regridding
targets_gcm = xr.Dataset(
    {
        "lat": (["lat"], np.arange(-90, 90 + gcm_res, gcm_res), {"units": "degrees_north"}),
        "lon": (["lon"], np.arange(-180, 180 + gcm_res, gcm_res), {"units": "degrees_east"}),
    }
)
regridder_gcm = xe.Regridder(gcm, targets_gcm, method='bilinear', periodic=False)
gcm = regridder_gcm(gcm[var])
# crop to CONUS
gcm = gcm.sel(lat=slice(lat_min, lat_max), lon=slice(lon_min, lon_max))
# fix datetime
dates = pd.date_range(gcm_start, gcm_end, freq='D')
# Remove leap days (Feb 29)
dates_no_leap = dates[~((dates.month == 2) & (dates.day == 29))]
gcm['time'] = dates_no_leap

# rcm is in rotated axis
rcm = xr.open_mfdataset(f'/projects/sds-lab/Shuochen/downscaling/cordex_canesm2/{var}.*.{gcm_name}.{rcm_name}.day.NAM-44.raw.nc')
# fix datetime
dates = pd.date_range(rcm_start, rcm_end, freq='D')
# Remove leap days (Feb 29)
dates_no_leap = dates[~((dates.month == 2) & (dates.day == 29))]
rcm['time'] = dates_no_leap
# Use a mask to crop to CONUS
lat = rcm['lat']
lon = rcm['lon']
mask = (lat >= 24) & (lat <= 50) & (lon >= -125) & (lon <= -66)
mask = mask.compute()
rcm = rcm.where(mask, drop=True)
# regridding
targets_rcm = xr.Dataset(
    {
        "lat": (["lat"], np.linspace(lat_min-offset,lat_max+offset,scaling_factor*len(gcm.lat)), {"units": "degrees_north"}),
        "lon": (["lon"], np.linspace(lon_min-offset,lon_max+offset,scaling_factor*len(gcm.lon)), {"units": "degrees_east"}),
    }
)
regridder_rcm = xe.Regridder(rcm, targets_rcm, method='bilinear', periodic=False)
rcm = regridder_rcm(rcm[var])

# align both datasets' time coord
gcm = gcm.sel(time=rcm.time)
# use PRISM elevation data to mask ocean and lake
elevation_regrid_2 = xr.open_dataset(prism_elevation)['elevation'].interp(lat=np.arange(lat_min,lat_max+gcm_res,gcm_res),lon=np.arange(lon_min,lon_max+gcm_res,gcm_res))
elevation_regrid_05 = xr.open_dataset(prism_elevation)['elevation'].interp(lat=np.linspace(lat_min-offset,lat_max+offset,scaling_factor*len(gcm.lat)),lon=np.linspace(lon_min-offset,lon_max+offset,scaling_factor*len(gcm.lon)))
gcm = gcm.where(~elevation_regrid_2.isnull())
rcm = rcm.where(~elevation_regrid_05.isnull())
# # export data
# elevation_regrid_2.to_netcdf("/projects/sds-lab/Shuochen/downscaling/elevation_regrid_2_conus.nc")
# elevation_regrid_05.to_netcdf("/projects/sds-lab/Shuochen/downscaling/elevation_regrid_05_conus.nc")
# gcm.to_netcdf("/projects/sds-lab/Shuochen/downscaling/canesm2/gcm_2deg_conus.nc")
# rcm.to_netcdf("/projects/sds-lab/Shuochen/downscaling/cordex_canesm2/rcm_0.5deg_conus.nc")
```
