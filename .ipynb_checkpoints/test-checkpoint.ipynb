```python
# =========================================================
# Notebook test code for SRResNet / SRResNet_HR_Aux
# Compatible with the current training script
# =========================================================

import os
import glob
import json

import torch
import numpy as np
import matplotlib.pyplot as plt
import torch.nn.functional as F

from torch import nn
from torch.utils.data import DataLoader, TensorDataset

from models import SRResNet, SRResNet_HR_Aux


# =========================================================
# Optional Cartopy imports
# =========================================================
try:
    import xarray as xr
    import cartopy.crs as ccrs
    import cartopy.feature as cfeature
    from mpl_toolkits.axes_grid1 import make_axes_locatable
    HAS_CARTOPY = True
except Exception as e:
    print("Cartopy/xarray plotting unavailable. Will use imshow instead.")
    print("Reason:", repr(e))
    HAS_CARTOPY = False


# =========================================================
# User settings: modify these in notebook
# =========================================================
rcm_var = "tmean"          # "tmean", "tmax", "tmin", or "prec"
gcm_name = "CanESM2"
rcm_name = "CanRCM4"
grid = "NAM-44i"
factor = 4

# Must match the training run
INPUT_SOURCE = "gcm"       # "gcm" or "coarse"
model_name = "SRResNet"    # "SRResNet" or "SRResNet_HR_Aux"

# Choose checkpoint
TRIAL_NUMBER = 236
USE_BEST = False
CKPT_PATH = None           # set direct path if needed, otherwise keep None

# Testing options
BATCH_SIZE = 1024
RANDOM_SEED = 42
EVAL_SPLIT = "val"         # "train", "val", "gap", or "all"
SAMPLE_INDEX = 0           # sample inside selected split for plotting

# For SRResNet_HR_Aux
HR_AUX_CHANNELS = 2
HR_ELEVATION_FILE = "high_res_elevation.pth"

# Plot options
NO_PLOT = False
SAVE_PLOT = False
PLOT_PATH = None

# Region used in preprocessing
lat_min, lat_max = 24.0, 50.0
lon_min, lon_max = -125.0, -66.0

# Optional coordinate file pattern. Keep None to guess automatically.
COORD_NC_PATTERN = None


# =========================================================
# Derived settings
# =========================================================
if model_name == "SRResNet_HR_Aux":
    INPUT_CHANNELS = 1
else:
    INPUT_CHANNELS = 3   # LR climate + LR mask + LR elevation

exp_folder_name = (
    f"/projects/sds-lab/Shuochen/downscaling/CORDEX/"
    f"{rcm_var}.{gcm_name}.{rcm_name}.day.{grid}.GCM_to_HR_RCM/"
)

if INPUT_SOURCE == "gcm":
    input_file = "low_res.pth"
    input_mask_file = "low_res_mask.pth"
    input_elev_file = "low_res_elevation.pth"
    input_tag = "GCM_low_res"

elif INPUT_SOURCE == "coarse":
    input_file = f"coarse_{factor}x.pth"
    input_mask_file = f"coarse_{factor}x_mask.pth"
    input_elev_file = f"coarse_{factor}x_elevation.pth"
    input_tag = f"coarsened_RCM_{factor}x"

else:
    raise ValueError(f"Unsupported INPUT_SOURCE: {INPUT_SOURCE}")

coarse_target_file = f"coarse_{factor}x.pth"
coarse_target_mask_file = f"coarse_{factor}x_mask.pth"

if model_name == "SRResNet_HR_Aux":
    input_setup_tag = f"{input_tag}_with_HR_mask_elevation_aux"
else:
    input_setup_tag = f"{input_tag}_with_mask_elevation"

save_root = os.path.join(
    exp_folder_name,
    "trained_models",
    model_name,
    input_setup_tag,
)

study_summary_path = os.path.join(save_root, "optuna_best_params.json")

print("Experiment folder:", exp_folder_name)
print("Save root:", save_root)
print("Input file:", input_file)
print("Input mask file:", input_mask_file)
print("Input elevation file:", input_elev_file)
print("Model:", model_name)
print("Input source:", INPUT_SOURCE)

# =========================================================
# Helper functions
# =========================================================

def match_sample_dim(tensor, n_samples, name):
    """
    Allows either:
      [N, C, H, W] full-time tensor
    or:
      [1, C, H, W] static tensor.
    """
    if tensor.shape[0] == n_samples:
        return tensor

    if tensor.shape[0] == 1:
        print(f"{name} has one sample; expanding to {n_samples} samples.")
        return tensor.expand(n_samples, -1, -1, -1)

    raise ValueError(
        f"{name} has incompatible sample dimension: "
        f"{tensor.shape[0]} vs expected {n_samples}"
    )


def masked_mse_sum(pred, target, mask):
    se = (pred - target) ** 2
    se = se * mask
    return se.sum().item(), mask.sum().item()


def get_year_split_indices(gcm_name):
    """
    Same year split as training script.
    """
    if gcm_name == "CanESM2":
        start_year = 1950
        train_end_year = 2005
        val_start_year = 2081

        train_end_idx = (train_end_year - start_year + 1) * 365
        val_start_idx = (val_start_year - start_year) * 365

    elif gcm_name == "EC-EARTH":
        def is_leap(year):
            return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

        def days_between_years(start_year, end_year_inclusive):
            total = 0
            for y in range(start_year, end_year_inclusive + 1):
                total += 366 if is_leap(y) else 365
            return total

        start_year = 1951
        train_end_year = 2005
        val_start_year = 2081

        train_end_idx = days_between_years(start_year, train_end_year)
        val_start_idx = days_between_years(start_year, val_start_year - 1)

    else:
        raise ValueError(f"Year split not defined for gcm_name={gcm_name}")

    return start_year, train_end_year, val_start_year, train_end_idx, val_start_idx


def as_cpu_float_tensor(x):
    if isinstance(x, torch.Tensor):
        return x.detach().cpu().float()
    return torch.tensor(x, dtype=torch.float32)


def denorm_y(y_pred_n, y_mean, y_std):
    return y_pred_n * y_std + y_mean


def build_model_from_checkpoint(ckpt, model_name, factor):
    h = ckpt["model_hparams"]

    if model_name == "SRResNet":
        model = SRResNet(
            num_resblk=h["num_resblk"],
            num_features=h["num_features"],
            input_channels=h.get("input_channels", INPUT_CHANNELS),
            output_channels=h.get("output_channels", 1),
            scale=h.get("scale", factor),
        )

    elif model_name == "SRResNet_HR_Aux":
        model = SRResNet_HR_Aux(
            num_resblk=h["num_resblk"],
            num_features=h["num_features"],
            input_channels=h.get("input_channels", 1),
            output_channels=h.get("output_channels", 1),
            hr_aux_channels=h.get("hr_aux_channels", HR_AUX_CHANNELS),
            scale=h.get("scale", factor),
        )

    else:
        raise ValueError(f"Unsupported model_name: {model_name}")

    return model


def try_reconstruct_coords(field_shape):
    """
    Attempts to reconstruct HR lat/lon coordinates for Cartopy plotting.
    Falls back to imshow if it fails.
    """
    if not HAS_CARTOPY:
        return None

    if COORD_NC_PATTERN is not None:
        nc_pattern = COORD_NC_PATTERN
    else:
        nc_pattern = (
            f"/projects/sds-lab/Shuochen/downscaling/CORDEX/"
            f"{rcm_var}.*.{gcm_name}.{rcm_name}.day.{grid}*.nc"
        )

    matches = sorted(glob.glob(nc_pattern))

    if len(matches) == 0:
        print("No coordinate netCDF files found. Falling back to imshow.")
        print("Tried pattern:", nc_pattern)
        return None

    try:
        ds = xr.open_mfdataset(matches, combine="by_coords")
        da = ds[rcm_var]

        lat_name, lon_name = None, None

        for name in da.coords:
            low = name.lower()
            if low in ["lat", "latitude"]:
                lat_name = name
            elif low in ["lon", "longitude"]:
                lon_name = name

        if lat_name is None or lon_name is None:
            print("Could not find lat/lon coordinate names. Falling back to imshow.")
            return None

        if lat_name not in da.dims or lon_name not in da.dims:
            print("Lat/lon are not 1D dimensions. Falling back to imshow.")
            return None

        da = da.sel(
            {
                lat_name: slice(lat_min, lat_max),
                lon_name: slice(lon_min, lon_max),
            }
        )

        lat = np.asarray(da[lat_name].values)
        lon = np.asarray(da[lon_name].values)

        H, W = len(lat), len(lon)

        Hf = (H // factor) * factor
        Wf = (W // factor) * factor

        trim_lat = H - Hf
        trim_lon = W - Wf

        top = trim_lat // 2
        bottom = top + Hf
        left = trim_lon // 2
        right = left + Wf

        lat_trim = lat[top:bottom]
        lon_trim = lon[left:right]

        if field_shape != (len(lat_trim), len(lon_trim)):
            print("Coordinate shape does not match tensor shape. Falling back to imshow.")
            print("Field shape:", field_shape)
            print("Coord shape:", (len(lat_trim), len(lon_trim)))
            return None

        lon2d, lat2d = np.meshgrid(lon_trim, lat_trim)

        return {
            "lat_trim": lat_trim,
            "lon_trim": lon_trim,
            "lat2d": lat2d,
            "lon2d": lon2d,
        }

    except Exception as e:
        print("Coordinate reconstruction failed. Falling back to imshow.")
        print("Reason:", repr(e))
        return None

# =========================================================
# Load checkpoint
# =========================================================

if CKPT_PATH is not None:
    ckpt_path = CKPT_PATH

elif USE_BEST:
    if not os.path.exists(study_summary_path):
        raise FileNotFoundError(
            f"Could not find optuna_best_params.json:\n{study_summary_path}"
        )

    with open(study_summary_path, "r") as f:
        best_summary = json.load(f)

    ckpt_path = best_summary.get("best_checkpoint_path", None)

    if ckpt_path is None:
        best_trial_number = best_summary["best_trial_number"]
        ckpt_path = os.path.join(save_root, f"trial_{best_trial_number:03d}.pth")

else:
    ckpt_path = os.path.join(save_root, f"trial_{TRIAL_NUMBER:03d}.pth")

if not os.path.exists(ckpt_path):
    raise FileNotFoundError(f"Checkpoint not found:\n{ckpt_path}")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

torch.manual_seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)

if torch.cuda.is_available():
    torch.cuda.manual_seed_all(RANDOM_SEED)

print("Loading checkpoint from:", ckpt_path)
ckpt = torch.load(ckpt_path, map_location=device)

print("\nLoaded checkpoint info:")
print("Trial number:", ckpt.get("trial_number", None))
print("Best epoch:", ckpt.get("best_epoch", None))
print("Best val loss:", ckpt.get("best_val_loss", None))
print("Checkpoint model_choice:", ckpt.get("model_choice", None))
print("Checkpoint input_source:", ckpt.get("input_source", None))
print("Model hparams:", ckpt.get("model_hparams", None))

if ckpt.get("model_choice", model_name) != model_name:
    raise ValueError(
        f"Checkpoint was trained with model_choice={ckpt.get('model_choice')}, "
        f"but notebook uses model_name={model_name}."
    )

if ckpt.get("input_source", INPUT_SOURCE) != INPUT_SOURCE:
    raise ValueError(
        f"Checkpoint was trained with input_source={ckpt.get('input_source')}, "
        f"but notebook uses INPUT_SOURCE={INPUT_SOURCE}."
    )

model = build_model_from_checkpoint(ckpt, model_name, factor).to(device)
model.load_state_dict(ckpt["model_state"])
model.eval()

print("Model loaded successfully.")

# =========================================================
# Load tensors
# =========================================================

X = torch.load(os.path.join(exp_folder_name, input_file))
X_mask = torch.load(os.path.join(exp_folder_name, input_mask_file))

y = torch.load(os.path.join(exp_folder_name, "high_res.pth"))
mask_hr = torch.load(os.path.join(exp_folder_name, "high_res_mask.pth"))

y_coarse = torch.load(os.path.join(exp_folder_name, coarse_target_file))
mask_coarse = torch.load(os.path.join(exp_folder_name, coarse_target_mask_file))

X_elev = None
X_hr_elev = None

if model_name == "SRResNet":
    X_elev = torch.load(os.path.join(exp_folder_name, input_elev_file))

elif model_name == "SRResNet_HR_Aux":
    X_hr_elev = torch.load(os.path.join(exp_folder_name, HR_ELEVATION_FILE))

print("Loaded raw shapes:")
print("X          :", X.shape)
print("X_mask     :", X_mask.shape)

if X_elev is not None:
    print("X_elev     :", X_elev.shape)

if X_hr_elev is not None:
    print("X_hr_elev  :", X_hr_elev.shape)

print("y          :", y.shape)
print("mask_hr    :", mask_hr.shape)
print("y_coarse   :", y_coarse.shape)
print("mask_coarse:", mask_coarse.shape)


# =========================================================
# Shape checks and static/full expansion
# =========================================================

X_mask = match_sample_dim(X_mask, X.shape[0], "X_mask")
mask_hr = match_sample_dim(mask_hr, y.shape[0], "mask_hr")
y_coarse = match_sample_dim(y_coarse, X.shape[0], "y_coarse")
mask_coarse = match_sample_dim(mask_coarse, X.shape[0], "mask_coarse")

if X_elev is not None:
    X_elev = match_sample_dim(X_elev, X.shape[0], "X_elev")

if X_hr_elev is not None:
    X_hr_elev = match_sample_dim(X_hr_elev, y.shape[0], "X_hr_elev")

if X.shape != X_mask.shape:
    raise ValueError(f"X shape {X.shape} does not match X_mask shape {X_mask.shape}")

if y.shape != mask_hr.shape:
    raise ValueError(f"y shape {y.shape} does not match mask_hr shape {mask_hr.shape}")

if y_coarse.shape != mask_coarse.shape:
    raise ValueError(
        f"y_coarse shape {y_coarse.shape} does not match mask_coarse shape {mask_coarse.shape}"
    )

if y_coarse.shape != X.shape:
    raise ValueError(
        f"y_coarse shape {y_coarse.shape} must match LR input shape {X.shape}"
    )

if X_elev is not None and X_elev.shape != X.shape:
    raise ValueError(f"X_elev shape {X_elev.shape} does not match X shape {X.shape}")

if X_hr_elev is not None and X_hr_elev.shape != y.shape:
    raise ValueError(f"X_hr_elev shape {X_hr_elev.shape} does not match y shape {y.shape}")

if X.shape[0] != y.shape[0]:
    raise ValueError(f"X has {X.shape[0]} samples but y has {y.shape[0]} samples")

if y.shape[-2] != factor * X.shape[-2]:
    raise ValueError(
        f"HR height {y.shape[-2]} != {factor} * LR height {X.shape[-2]}"
    )

if y.shape[-1] != factor * X.shape[-1]:
    raise ValueError(
        f"HR width {y.shape[-1]} != {factor} * LR width {X.shape[-1]}"
    )


# =========================================================
# Dtype and mask formatting
# =========================================================

X = X.float()
y = y.float()
y_coarse = y_coarse.float()

X_mask = X_mask.float()
X_mask = (X_mask > 0.5).float()

mask_hr = mask_hr.float()
mask_hr = (mask_hr > 0.5).float()

mask_coarse = mask_coarse.float()
mask_coarse = (mask_coarse > 0.5).float()

if X_elev is not None:
    X_elev = X_elev.float()
    X_elev = torch.nan_to_num(X_elev, nan=0.0, posinf=0.0, neginf=0.0)

if X_hr_elev is not None:
    X_hr_elev = X_hr_elev.float()
    X_hr_elev = torch.nan_to_num(X_hr_elev, nan=0.0, posinf=0.0, neginf=0.0)

print("Tensor checks passed.")

# =========================================================
# Split data exactly like training
# =========================================================

(
    start_year,
    train_end_year,
    val_start_year,
    train_end_idx,
    val_start_idx,
) = get_year_split_indices(gcm_name)

print("Year split:")
print("start_year     :", start_year)
print("train_end_year :", train_end_year)
print("val_start_year :", val_start_year)
print("train_end_idx  :", train_end_idx)
print("val_start_idx  :", val_start_idx)
print("total samples  :", X.shape[0])

if EVAL_SPLIT == "train":
    eval_slice = slice(0, train_end_idx)
    eval_start_idx = 0

elif EVAL_SPLIT == "val":
    eval_slice = slice(val_start_idx, None)
    eval_start_idx = val_start_idx

elif EVAL_SPLIT == "gap":
    eval_slice = slice(train_end_idx, val_start_idx)
    eval_start_idx = train_end_idx

elif EVAL_SPLIT == "all":
    eval_slice = slice(0, None)
    eval_start_idx = 0

else:
    raise ValueError(f"Unsupported EVAL_SPLIT: {EVAL_SPLIT}")


X_eval = X[eval_slice]
X_mask_eval = X_mask[eval_slice]

y_eval = y[eval_slice]
mask_eval = mask_hr[eval_slice]

y_coarse_eval = y_coarse[eval_slice]
mask_coarse_eval = mask_coarse[eval_slice]

if X_elev is not None:
    X_elev_eval = X_elev[eval_slice]
else:
    X_elev_eval = None

if X_hr_elev is not None:
    X_hr_elev_eval = X_hr_elev[eval_slice]
else:
    X_hr_elev_eval = None

if X_eval.shape[0] == 0:
    raise ValueError(f"Selected EVAL_SPLIT={EVAL_SPLIT} has zero samples.")

print(f"Selected EVAL_SPLIT: {EVAL_SPLIT}")
print("X_eval          :", X_eval.shape)
print("X_mask_eval     :", X_mask_eval.shape)

if X_elev_eval is not None:
    print("X_elev_eval     :", X_elev_eval.shape)

if X_hr_elev_eval is not None:
    print("X_hr_elev_eval  :", X_hr_elev_eval.shape)

print("y_eval          :", y_eval.shape)
print("mask_eval       :", mask_eval.shape)
print("y_coarse_eval   :", y_coarse_eval.shape)
print("mask_coarse_eval:", mask_coarse_eval.shape)

# =========================================================
# Normalize using checkpoint stats and build model inputs
# =========================================================

X_mean = as_cpu_float_tensor(ckpt["X_mean"])
X_std = as_cpu_float_tensor(ckpt["X_std"])

y_mean = as_cpu_float_tensor(ckpt["y_mean"])
y_std = as_cpu_float_tensor(ckpt["y_std"])

if X_std.item() == 0:
    X_std = torch.tensor(1.0, dtype=torch.float32)

if y_std.item() == 0:
    y_std = torch.tensor(1.0, dtype=torch.float32)

print("Checkpoint normalization:")
print("X_mean:", X_mean.item())
print("X_std :", X_std.item())
print("y_mean:", y_mean.item())
print("y_std :", y_std.item())


# Normalize only climate input channel
X_eval_clim_n = (X_eval - X_mean) / X_std

# Force invalid LR pixels to zero after normalization
X_eval_clim_n = X_eval_clim_n * X_mask_eval

if model_name == "SRResNet":
    # channel 0: normalized LR climate
    # channel 1: LR land-sea mask
    # channel 2: LR elevation
    X_eval_in = torch.cat(
        [X_eval_clim_n, X_mask_eval, X_elev_eval],
        dim=1,
    ).contiguous()

    X_eval_hr_aux = None

elif model_name == "SRResNet_HR_Aux":
    # main input: normalized LR climate only
    X_eval_in = X_eval_clim_n.contiguous()

    # HR aux: HR land-sea mask + HR elevation
    X_eval_hr_aux = torch.cat(
        [mask_eval, X_hr_elev_eval],
        dim=1,
    ).contiguous()

else:
    raise ValueError(f"Unsupported model_name: {model_name}")


print("Model input shapes:")
print("X_eval_in:", X_eval_in.shape)

if X_eval_hr_aux is not None:
    print("X_eval_hr_aux:", X_eval_hr_aux.shape)

if X_eval_in.shape[1] != INPUT_CHANNELS:
    raise ValueError(
        f"Expected {INPUT_CHANNELS} input channels, got {X_eval_in.shape[1]}"
    )

if model_name == "SRResNet_HR_Aux":
    if X_eval_hr_aux.shape[1] != HR_AUX_CHANNELS:
        raise ValueError(
            f"Expected {HR_AUX_CHANNELS} HR auxiliary channels, "
            f"got {X_eval_hr_aux.shape[1]}"
        )

# =========================================================
# Evaluate model
# =========================================================

if model_name == "SRResNet":
    eval_set = TensorDataset(
        X_eval_in,
        y_eval,
        mask_eval,
        y_coarse_eval,
        mask_coarse_eval,
    )

elif model_name == "SRResNet_HR_Aux":
    eval_set = TensorDataset(
        X_eval_in,
        X_eval_hr_aux,
        y_eval,
        mask_eval,
        y_coarse_eval,
        mask_coarse_eval,
    )

eval_loader = DataLoader(
    eval_set,
    batch_size=BATCH_SIZE,
    shuffle=False,
)

y_mean_dev = y_mean.to(device)
y_std_dev = y_std.to(device)

total_sse_full = 0.0
total_numel_full = 0

total_sse_land = 0.0
total_land_count = 0.0

total_sse_coarse_land = 0.0
total_coarse_land_count = 0.0

model.eval()

with torch.no_grad():
    if model_name == "SRResNet":
        for Xn, y_raw, mask_raw, y_coarse_raw, mask_coarse_raw in eval_loader:
            Xn = Xn.to(device)
            y_raw = y_raw.to(device)
            mask_raw = mask_raw.to(device)

            y_coarse_raw = y_coarse_raw.to(device)
            mask_coarse_raw = mask_coarse_raw.to(device)

            y_pred_n = model(Xn)
            y_pred = denorm_y(y_pred_n, y_mean_dev, y_std_dev)

            # Ocean exactly zero in physical space
            y_pred = y_pred * mask_raw

            se_full = (y_pred - y_raw) ** 2
            total_sse_full += se_full.sum().item()
            total_numel_full += se_full.numel()

            sse_land, count_land = masked_mse_sum(y_pred, y_raw, mask_raw)
            total_sse_land += sse_land
            total_land_count += count_land

            y_pred_coarse = F.interpolate(
                y_pred,
                size=y_coarse_raw.shape[-2:],
                mode="bilinear",
                align_corners=False,
            )

            sse_coarse, count_coarse = masked_mse_sum(
                y_pred_coarse,
                y_coarse_raw,
                mask_coarse_raw,
            )
            total_sse_coarse_land += sse_coarse
            total_coarse_land_count += count_coarse

    elif model_name == "SRResNet_HR_Aux":
        for Xn, X_hr_aux, y_raw, mask_raw, y_coarse_raw, mask_coarse_raw in eval_loader:
            Xn = Xn.to(device)
            X_hr_aux = X_hr_aux.to(device)

            y_raw = y_raw.to(device)
            mask_raw = mask_raw.to(device)

            y_coarse_raw = y_coarse_raw.to(device)
            mask_coarse_raw = mask_coarse_raw.to(device)

            y_pred_n = model(Xn, X_hr_aux)
            y_pred = denorm_y(y_pred_n, y_mean_dev, y_std_dev)

            # Ocean exactly zero in physical space
            y_pred = y_pred * mask_raw

            se_full = (y_pred - y_raw) ** 2
            total_sse_full += se_full.sum().item()
            total_numel_full += se_full.numel()

            sse_land, count_land = masked_mse_sum(y_pred, y_raw, mask_raw)
            total_sse_land += sse_land
            total_land_count += count_land

            y_pred_coarse = F.interpolate(
                y_pred,
                size=y_coarse_raw.shape[-2:],
                mode="bilinear",
                align_corners=False,
            )

            sse_coarse, count_coarse = masked_mse_sum(
                y_pred_coarse,
                y_coarse_raw,
                mask_coarse_raw,
            )
            total_sse_coarse_land += sse_coarse
            total_coarse_land_count += count_coarse


full_image_mse = total_sse_full / total_numel_full
land_mse = total_sse_land / max(total_land_count, 1.0)
coarse_land_mse = total_sse_coarse_land / max(total_coarse_land_count, 1.0)

print("\n=========================================================")
print("Evaluation results")
print("=========================================================")
print("Model name        :", model_name)
print("Input source      :", INPUT_SOURCE)
print("Eval split        :", EVAL_SPLIT)
print("Checkpoint        :", ckpt_path)
print("Full-image MSE    :", f"{full_image_mse:.8f}")
print("HR land-only MSE  :", f"{land_mse:.8f}")
print("Coarse land MSE   :", f"{coarse_land_mse:.8f}")
print("=========================================================")


# =========================================================
# Plot one sample
# =========================================================

if not NO_PLOT:
    if SAMPLE_INDEX < 0 or SAMPLE_INDEX >= X_eval_in.shape[0]:
        raise ValueError(
            f"SAMPLE_INDEX must be in [0, {X_eval_in.shape[0] - 1}], "
            f"got {SAMPLE_INDEX}"
        )

    idx = SAMPLE_INDEX
    absolute_idx = eval_start_idx + idx

    with torch.no_grad():
        if model_name == "SRResNet":
            X_one = X_eval_in[idx:idx + 1].to(device)
            y_one = y_eval[idx:idx + 1].to(device)
            mask_one = mask_eval[idx:idx + 1].to(device)

            y_pred_n = model(X_one)

        elif model_name == "SRResNet_HR_Aux":
            X_one = X_eval_in[idx:idx + 1].to(device)
            X_aux_one = X_eval_hr_aux[idx:idx + 1].to(device)

            y_one = y_eval[idx:idx + 1].to(device)
            mask_one = mask_eval[idx:idx + 1].to(device)

            y_pred_n = model(X_one, X_aux_one)

        y_pred = denorm_y(y_pred_n, y_mean_dev, y_std_dev)
        y_pred = y_pred * mask_one

    f_true = y_one[0, 0].detach().cpu().numpy()
    f_pred = y_pred[0, 0].detach().cpu().numpy()

    diff_pred = f_true - f_pred
    abs_err_pred = np.abs(diff_pred)

    vmin = np.nanmin(f_true)
    vmax = np.nanmax(f_true)

    diff_lim = np.nanmax(np.abs(diff_pred))
    abs_lim = np.nanmax(abs_err_pred)

    coords = try_reconstruct_coords(f_true.shape)

    if coords is not None:
        lat_trim = coords["lat_trim"]
        lon_trim = coords["lon_trim"]
        lat2d = coords["lat2d"]
        lon2d = coords["lon2d"]

        if lat_trim[0] > lat_trim[-1]:
            f_true = np.flipud(f_true)
            f_pred = np.flipud(f_pred)
            diff_pred = np.flipud(diff_pred)
            abs_err_pred = np.flipud(abs_err_pred)

            lat_trim = lat_trim[::-1]
            lon2d, lat2d = np.meshgrid(lon_trim, lat_trim)

        def plot_map(ax, data, title, vmin=None, vmax=None, cmap="viridis"):
            im = ax.pcolormesh(
                lon2d,
                lat2d,
                data,
                transform=ccrs.PlateCarree(),
                cmap=cmap,
                vmin=vmin,
                vmax=vmax,
                shading="auto",
            )

            ax.set_extent(
                [lon_trim.min(), lon_trim.max(), lat_trim.min(), lat_trim.max()],
                crs=ccrs.PlateCarree(),
            )

            ax.coastlines(resolution="110m", linewidth=0.8)
            ax.add_feature(cfeature.BORDERS, linewidth=0.5)
            ax.add_feature(cfeature.STATES, linewidth=0.3)
            ax.set_title(title, fontsize=11)

            return im

        def add_vertical_cbar(ax, im):
            divider = make_axes_locatable(ax)
            cax = divider.append_axes(
                "right",
                size="3%",
                pad=0.1,
                axes_class=plt.Axes,
            )
            return plt.colorbar(im, cax=cax)

        proj = ccrs.PlateCarree()
        fig = plt.figure(figsize=(12, 7.2))
        gs = fig.add_gridspec(2, 2, hspace=0.08, wspace=0.08)

        axes = [
            fig.add_subplot(gs[0, 0], projection=proj),
            fig.add_subplot(gs[0, 1], projection=proj),
            fig.add_subplot(gs[1, 0], projection=proj),
            fig.add_subplot(gs[1, 1], projection=proj),
        ]

        im0 = plot_map(
            axes[0],
            f_true,
            "High-res RCM truth",
            vmin=vmin,
            vmax=vmax,
            cmap="Reds",
        )

        im1 = plot_map(
            axes[1],
            f_pred,
            f"{model_name} prediction\nFull MSE={full_image_mse:.6f}",
            vmin=vmin,
            vmax=vmax,
            cmap="Reds",
        )

        im2 = plot_map(
            axes[2],
            diff_pred,
            "Difference: truth - prediction",
            vmin=-diff_lim,
            vmax=diff_lim,
            cmap="RdBu_r",
        )

        im3 = plot_map(
            axes[3],
            abs_err_pred,
            "|Error|",
            vmin=0,
            vmax=abs_lim,
            cmap="Reds",
        )

        for ax, im in zip(axes, [im0, im1, im2, im3]):
            add_vertical_cbar(ax, im)

        fig.suptitle(
            (
                f"{rcm_var} | {gcm_name}-{rcm_name} | {EVAL_SPLIT} split | "
                f"sample={idx}, absolute_index={absolute_idx}"
            ),
            fontsize=13,
        )

        plt.subplots_adjust(left=0.04, right=0.97, top=0.92, bottom=0.06)

    else:
        fig, axes = plt.subplots(2, 2, figsize=(12, 7.2))

        im0 = axes[0, 0].imshow(f_true, vmin=vmin, vmax=vmax, cmap="Reds")
        axes[0, 0].set_title("High-res RCM truth")

        im1 = axes[0, 1].imshow(f_pred, vmin=vmin, vmax=vmax, cmap="Reds")
        axes[0, 1].set_title(f"{model_name} prediction\nFull MSE={full_image_mse:.6f}")

        im2 = axes[1, 0].imshow(
            diff_pred,
            vmin=-diff_lim,
            vmax=diff_lim,
            cmap="RdBu_r",
        )
        axes[1, 0].set_title("Difference: truth - prediction")

        im3 = axes[1, 1].imshow(
            abs_err_pred,
            vmin=0,
            vmax=abs_lim,
            cmap="Reds",
        )
        axes[1, 1].set_title("|Error|")

        for ax, im in zip(axes.ravel(), [im0, im1, im2, im3]):
            plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
            ax.set_xticks([])
            ax.set_yticks([])

        fig.suptitle(
            (
                f"{rcm_var} | {gcm_name}-{rcm_name} | {EVAL_SPLIT} split | "
                f"sample={idx}, absolute_index={absolute_idx}"
            ),
            fontsize=13,
        )

        plt.tight_layout()

    if SAVE_PLOT:
        if PLOT_PATH is None:
            plot_path = os.path.join(
                save_root,
                f"test_plot_{model_name}_{INPUT_SOURCE}_{EVAL_SPLIT}_sample{idx}.png",
            )
        else:
            plot_path = PLOT_PATH

        os.makedirs(os.path.dirname(plot_path), exist_ok=True)
        plt.savefig(plot_path, dpi=300, bbox_inches="tight")
        print("Saved plot to:", plot_path)

    plt.show()
```
