```python
N_TRIALS = 30
EPOCHS = 15
EARLY_STOPPING_PATIENCE = 30
MIN_DELTA = 0.0
RANDOM_SEED = 42

# =========================================================
# experiment settings
# =========================================================
rcm_var = "tmean"   # 'tmean' or 'prec'
gcm_name = "CanESM2"
gcm_rcm_dict = {"CanESM2": "CanRCM4"}
grid = "NAM-44i"
factor = 4

exp_folder_name = (
    f"/projects/sds-lab/Shuochen/downscaling/cordex_canesm2/"
    f"{rcm_var}.{gcm_name}.{gcm_rcm_dict[gcm_name]}.day.{grid}.mbcn-gridMET/"
)

# =========================================================
# explicit year-based split
# source/train = past climate
# target/val   = future climate
# no additional split inside either set
# assumes 365 days per year (no leap years)
# =========================================================
start_year = 1950
train_end_year = 2005
val_start_year = 2081
end_year = 2100

import os
import copy
import json
import math
import torch
import optuna
import numpy as np
from itertools import cycle
from torch import nn
from torch.utils.data import DataLoader, TensorDataset
from torch.optim.lr_scheduler import ReduceLROnPlateau

from models import SRResNet_DANN

# =========================================================
# device
# =========================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

torch.manual_seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(RANDOM_SEED)

# =========================================================
# save paths
# =========================================================
model_name = "SRResNet_DANN"
save_root = os.path.join(exp_folder_name, "trained_models", model_name)
os.makedirs(save_root, exist_ok=True)

study_summary_path = os.path.join(save_root, "optuna_best_params.json")
all_trials_summary_path = os.path.join(save_root, "optuna_all_trials_summary.json")

print("All trial checkpoints will be saved to:", save_root)

# =========================================================
# load tensors
# =========================================================
X = torch.load(os.path.join(exp_folder_name, f"coarse_{factor}x.pth"))
y = torch.load(os.path.join(exp_folder_name, "high_res.pth"))
mask_hr = torch.load(os.path.join(exp_folder_name, "high_res_mask.pth"))

print("Loaded shapes:")
print("X       :", X.shape)
print("y       :", y.shape)
print("mask_hr :", mask_hr.shape)

if y.shape != mask_hr.shape:
    raise ValueError(
        f"Shape mismatch: y shape {y.shape} does not match mask shape {mask_hr.shape}"
    )

mask_hr = mask_hr.float()
mask_hr = (mask_hr > 0.5).float()

# =========================================================
# explicit temporal split
# source/train = start_year ... train_end_year
# target/val   = val_start_year ... end_year
# years in between are skipped
# =========================================================
if train_end_year < start_year:
    raise ValueError("train_end_year must be >= start_year.")
if val_start_year > end_year:
    raise ValueError("val_start_year must be <= end_year.")
if val_start_year <= train_end_year:
    raise ValueError("val_start_year must be greater than train_end_year.")

train_end_idx = (train_end_year - start_year + 1) * 365
val_start_idx = (val_start_year - start_year) * 365
expected_total = (end_year - start_year + 1) * 365

if X.shape[0] != expected_total:
    raise ValueError(
        f"Expected {expected_total} samples for {start_year}-{end_year} "
        f"with no leap years, but got {X.shape[0]}"
    )

if train_end_idx <= 0 or train_end_idx > X.shape[0]:
    raise ValueError("Computed train_end_idx is out of range.")
if val_start_idx < 0 or val_start_idx >= X.shape[0]:
    raise ValueError("Computed val_start_idx is out of range.")

X_source_train = X[:train_end_idx]
y_source_train = y[:train_end_idx]
mask_source_train = mask_hr[:train_end_idx]

X_target_val = X[val_start_idx:]
y_target_val = y[val_start_idx:]
mask_target_val = mask_hr[val_start_idx:]

print("\nYear-based split:")
print(f"Source train : {start_year}-01-01 to {train_end_year}-12-31")
print(f"Skipped years: {train_end_year + 1}-01-01 to {val_start_year - 1}-12-31")
print(f"Target val   : {val_start_year}-01-01 to {end_year}-12-31")

print("\nSource train shapes:")
print("X_source_train    :", X_source_train.shape)
print("y_source_train    :", y_source_train.shape)
print("mask_source_train :", mask_source_train.shape)

print("\nTarget val shapes:")
print("X_target_val      :", X_target_val.shape)
print("y_target_val      :", y_target_val.shape)
print("mask_target_val   :", mask_target_val.shape)

# =========================================================
# normalization stats from SOURCE TRAIN only
# =========================================================
X_mean = X_source_train.mean()
X_std = X_source_train.std()
X_max = X_source_train.max()

y_source_train_land = y_source_train[mask_source_train > 0.5]
if y_source_train_land.numel() == 0:
    raise ValueError("No land points found in source training mask.")

y_mean = y_source_train_land.mean()
y_std = y_source_train_land.std()
y_max = y_source_train_land.max()

if X_std.item() == 0:
    X_std = torch.tensor(1.0, dtype=X_source_train.dtype)
if y_std.item() == 0:
    y_std = torch.tensor(1.0, dtype=y_source_train.dtype)

# normalize source with source stats
X_source_train_n = (X_source_train - X_mean) / X_std
y_source_train_n = (y_source_train - y_mean) / y_std

# normalize target with SOURCE stats
X_target_val_n = (X_target_val - X_mean) / X_std

# =========================================================
# losses
# =========================================================
def masked_mse(pred, target, mask, eps=1e-8):
    """
    pred, target, mask: [B, C, H, W]
    mask = 1 over land, 0 over ocean
    Returns mean squared error over land pixels only.
    """
    se = (pred - target) ** 2
    se = se * mask
    return se.sum() / (mask.sum() + eps)

sr_loss_fn = masked_mse
val_loss_fn = nn.MSELoss()
domain_loss_fn = nn.CrossEntropyLoss()

# =========================================================
# datasets
# source_training_set: labeled source domain
# target_training_set: unlabeled target domain (same target period)
# target_validation_set: labeled target period for validation/reporting
# =========================================================
source_training_set = TensorDataset(
    X_source_train_n,
    y_source_train_n,
    X_source_train,
    y_source_train,
    mask_source_train
)

target_training_set = TensorDataset(
    X_target_val_n
)

target_validation_set = TensorDataset(
    X_target_val_n,
    y_target_val,
    mask_target_val
)

# =========================================================
# helper for DANN lambda schedule
# =========================================================
def dann_lambda_schedule(progress):
    """
    progress in [0, 1]
    standard DANN schedule
    """
    return 2.0 / (1.0 + math.exp(-10.0 * progress)) - 1.0

# =========================================================
# objective function
# =========================================================
def objective(trial):
    # -----------------------------------------------------
    # hyperparameters to search
    # -----------------------------------------------------
    batch_size = trial.suggest_categorical("batch_size", [256, 512, 1024])
    learning_rate = trial.suggest_float("learning_rate", 1e-4, 3e-4, log=True)
    num_resblk = trial.suggest_int("num_resblk", 4, 16, step=2)
    num_features = trial.suggest_categorical("num_features", [32, 64, 96, 128])
    weight_decay = trial.suggest_float("weight_decay", 1e-6, 1e-3, log=True)
    domain_hidden_dim = trial.suggest_categorical("domain_hidden_dim", [64, 128, 256])
    alpha_domain = trial.suggest_float("alpha_domain", 0.01, 1.0, log=True)

    # -----------------------------------------------------
    # paths for this trial
    # -----------------------------------------------------
    trial_ckpt_path = os.path.join(save_root, f"trial_{trial.number:03d}.pth")

    # -----------------------------------------------------
    # model
    # -----------------------------------------------------
    model = SRResNet_DANN(
        num_resblk=num_resblk,
        num_features=num_features,
        input_channels=1,
        output_channels=1,
        scale=factor,
        num_domains=2,
        domain_hidden_dim=domain_hidden_dim,
        lambda_grl=1.0,
    ).to(device)

    # -----------------------------------------------------
    # dataloaders
    # -----------------------------------------------------
    source_train_dataloader = DataLoader(
        source_training_set,
        batch_size=batch_size,
        shuffle=True,
        drop_last=True,
    )

    target_train_dataloader = DataLoader(
        target_training_set,
        batch_size=batch_size,
        shuffle=True,
        drop_last=True,
    )

    target_val_dataloader = DataLoader(
        target_validation_set,
        batch_size=batch_size,
        shuffle=False,
    )

    if len(source_train_dataloader) == 0:
        raise ValueError("Source train dataloader is empty. Reduce batch size or check data size.")
    if len(target_train_dataloader) == 0:
        raise ValueError("Target train dataloader is empty. Reduce batch size or check data size.")
    if len(target_val_dataloader) == 0:
        raise ValueError("Target validation dataloader is empty. Reduce batch size or check data size.")

    # -----------------------------------------------------
    # training setup
    # -----------------------------------------------------
    optimizer = torch.optim.AdamW(
        model.parameters(),
        lr=learning_rate,
        weight_decay=weight_decay,
    )
    scheduler = ReduceLROnPlateau(optimizer, patience=15, factor=0.5)

    train_total_loss_list = []
    train_sr_loss_list = []
    train_domain_loss_list = []
    target_val_loss_list = []

    best_val_loss = float("inf")
    best_epoch = -1
    epochs_no_improve = 0

    y_mean_dev = y_mean.to(device)
    y_std_dev = y_std.to(device)

    total_training_steps = EPOCHS * max(len(source_train_dataloader), len(target_train_dataloader))
    global_step = 0

    # -----------------------------------------------------
    # train loop
    # -----------------------------------------------------
    for epoch in range(EPOCHS):
        model.train()

        epoch_total_loss = 0.0
        epoch_sr_loss = 0.0
        epoch_domain_loss = 0.0
        num_batches = 0

        target_iter = cycle(target_train_dataloader)

        for Xs_n, ys_n, Xs_raw, ys_raw, mask_s_raw in source_train_dataloader:
            Xt_n = next(target_iter)[0]

            Xs_n = Xs_n.to(device)
            ys_n = ys_n.to(device)
            mask_s_raw = mask_s_raw.to(device)

            Xt_n = Xt_n.to(device)

            progress = global_step / max(total_training_steps - 1, 1)
            lambda_grl = dann_lambda_schedule(progress)

            # source forward
            ys_pred_n, domain_logits_s = model(Xs_n, lambda_grl=lambda_grl)

            # target forward
            _, domain_logits_t = model(Xt_n, lambda_grl=lambda_grl)

            # SR loss on source only
            ys_pred_n = ys_pred_n * mask_s_raw
            loss_sr = sr_loss_fn(ys_pred_n, ys_n, mask_s_raw)

            # domain loss
            domain_labels_s = torch.zeros(
                Xs_n.size(0), dtype=torch.long, device=device
            )  # 0 = source/past
            domain_labels_t = torch.ones(
                Xt_n.size(0), dtype=torch.long, device=device
            )  # 1 = target/future

            loss_domain_s = domain_loss_fn(domain_logits_s, domain_labels_s)
            loss_domain_t = domain_loss_fn(domain_logits_t, domain_labels_t)
            loss_domain = loss_domain_s + loss_domain_t

            # total loss
            loss = loss_sr + alpha_domain * loss_domain

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            epoch_total_loss += loss.item()
            epoch_sr_loss += loss_sr.item()
            epoch_domain_loss += loss_domain.item()
            num_batches += 1
            global_step += 1

        train_total_loss = epoch_total_loss / max(num_batches, 1)
        train_sr_loss = epoch_sr_loss / max(num_batches, 1)
        train_domain_loss = epoch_domain_loss / max(num_batches, 1)

        train_total_loss_list.append(train_total_loss)
        train_sr_loss_list.append(train_sr_loss)
        train_domain_loss_list.append(train_domain_loss)

        # -----------------------------------------------------
        # validation on target period
        # full-image MSE in physical space
        # -----------------------------------------------------
        model.eval()
        target_val_loss = 0.0

        with torch.no_grad():
            for Xt_n, yt_raw, mask_t_raw in target_val_dataloader:
                Xt_n = Xt_n.to(device)
                yt_raw = yt_raw.to(device)
                mask_t_raw = mask_t_raw.to(device)

                yt_pred_n, _ = model(Xt_n, lambda_grl=0.0)
                yt_pred = yt_pred_n * y_std_dev + y_mean_dev

                # exact zero over ocean
                yt_pred = yt_pred * mask_t_raw

                # full-image MSE
                target_val_loss += val_loss_fn(yt_pred, yt_raw).item()

        target_val_loss /= len(target_val_dataloader)
        target_val_loss_list.append(target_val_loss)

        scheduler.step(target_val_loss)

        # -----------------------------------------------------
        # report to Optuna using target validation only
        # -----------------------------------------------------
        trial.report(target_val_loss, epoch)

        if trial.should_prune():
            trial.set_user_attr("status", "pruned")
            raise optuna.TrialPruned()

        # -----------------------------------------------------
        # save best model using target validation only
        # -----------------------------------------------------
        if target_val_loss < best_val_loss - MIN_DELTA:
            best_val_loss = target_val_loss
            best_epoch = epoch
            epochs_no_improve = 0

            trial.set_user_attr("best_epoch", best_epoch)
            trial.set_user_attr("best_val_loss", best_val_loss)
            trial.set_user_attr("checkpoint_path", trial_ckpt_path)

            torch.save(
                {
                    "trial_number": trial.number,
                    "epoch": best_epoch,
                    "model_name": model.__class__.__name__,
                    "model_hparams": {
                        "num_resblk": num_resblk,
                        "num_features": num_features,
                        "input_channels": 1,
                        "output_channels": 1,
                        "scale": factor,
                        "num_domains": 2,
                        "domain_hidden_dim": domain_hidden_dim,
                    },
                    "training_hparams": {
                        "batch_size": batch_size,
                        "learning_rate": learning_rate,
                        "weight_decay": weight_decay,
                        "alpha_domain": alpha_domain,
                        "epochs": EPOCHS,
                        "early_stopping_patience": EARLY_STOPPING_PATIENCE,
                        "min_delta": MIN_DELTA,
                        "loss": "source_masked_mse + alpha_domain * domain_ce",
                        "validation_loss": "target_full_image_mse",
                        "ocean_constraint": "prediction multiplied by land mask in physical space",
                        "domain_definition": "source=past_climate_train_period, target=future_validation_period",
                        "start_year": start_year,
                        "train_end_year": train_end_year,
                        "val_start_year": val_start_year,
                        "end_year": end_year,
                        "note": "no additional split inside source or target; target validation period is also used as unlabeled target domain during DANN training",
                    },
                    "model_state": copy.deepcopy(model.state_dict()),
                    "optimizer_state": optimizer.state_dict(),
                    "scheduler_state": scheduler.state_dict(),
                    "best_val_loss": best_val_loss,
                    "best_epoch": best_epoch,
                    "target_val_loss": target_val_loss,
                    "train_total_loss": train_total_loss,
                    "train_sr_loss": train_sr_loss,
                    "train_domain_loss": train_domain_loss,
                    "train_total_loss_list": train_total_loss_list,
                    "train_sr_loss_list": train_sr_loss_list,
                    "train_domain_loss_list": train_domain_loss_list,
                    "target_val_loss_list": target_val_loss_list,
                    "X_mean": X_mean,
                    "X_std": X_std,
                    "X_max": X_max,
                    "y_mean": y_mean,
                    "y_std": y_std,
                    "y_max": y_max,
                },
                trial_ckpt_path,
            )
        else:
            epochs_no_improve += 1

        print(
            f"Trial {trial.number:03d} | "
            f"Epoch {epoch:03d} | "
            f"Train Total: {train_total_loss:.6f} | "
            f"Train SR: {train_sr_loss:.6f} | "
            f"Train Domain: {train_domain_loss:.6f} | "
            f"Target Val: {target_val_loss:.6f} | "
            f"Best Val: {best_val_loss:.6f} | "
            f"Best Epoch: {best_epoch:03d} | "
            f"lambda_grl: {lambda_grl:.4f} | "
            f"No improve: {epochs_no_improve}/{EARLY_STOPPING_PATIENCE} | "
            f"lr: {optimizer.param_groups[0]['lr']:.6e}"
        )

        if epochs_no_improve >= EARLY_STOPPING_PATIENCE:
            print(
                f"Early stopping triggered for trial {trial.number:03d} "
                f"at epoch {epoch:03d} | best epoch: {best_epoch:03d}"
            )
            break

    trial.set_user_attr("status", "completed")
    trial.set_user_attr("best_epoch", best_epoch)
    trial.set_user_attr("best_val_loss", best_val_loss)
    trial.set_user_attr("checkpoint_path", trial_ckpt_path)

    return best_val_loss


# =========================================================
# create and run study
# =========================================================
study = optuna.create_study(
    direction="minimize",
    pruner=optuna.pruners.MedianPruner(n_startup_trials=5, n_warmup_steps=5),
)

study.optimize(objective, n_trials=N_TRIALS)

# =========================================================
# save best params summary
# =========================================================
best_trial = study.best_trial
best_summary = {
    "best_trial_number": best_trial.number,
    "best_value": best_trial.value,
    "best_params": best_trial.params,
    "best_epoch": best_trial.user_attrs.get("best_epoch", None),
    "best_checkpoint_path": best_trial.user_attrs.get(
        "checkpoint_path",
        os.path.join(save_root, f"trial_{best_trial.number:03d}.pth")
    ),
}

with open(study_summary_path, "w") as f:
    json.dump(best_summary, f, indent=4)

# =========================================================
# save all trial summaries
# =========================================================
all_trials_summary = []
for t in study.trials:
    trial_info = {
        "trial_number": t.number,
        "state": str(t.state),
        "value": t.value if t.value is not None else None,
        "params": t.params,
        "best_epoch": t.user_attrs.get("best_epoch", None),
        "best_val_loss": t.user_attrs.get("best_val_loss", None),
        "checkpoint_path": t.user_attrs.get(
            "checkpoint_path",
            os.path.join(save_root, f"trial_{t.number:03d}.pth")
        ),
    }
    all_trials_summary.append(trial_info)

with open(all_trials_summary_path, "w") as f:
    json.dump(all_trials_summary, f, indent=4)

# =========================================================
# final logs
# =========================================================
print("\nOptuna study finished.")
print(f"Best trial: {best_trial.number}")
print(f"Best value (target val): {best_trial.value:.6f}")
print(f"Best epoch: {best_trial.user_attrs.get('best_epoch', None)}")
print("Best params:", best_trial.params)
print(
    "Best checkpoint:",
    best_trial.user_attrs.get(
        "checkpoint_path",
        os.path.join(save_root, f"trial_{best_trial.number:03d}.pth")
    )
)
print("Best summary saved to:", study_summary_path)
print("All trial summaries saved to:", all_trials_summary_path)
print("All per-trial checkpoints saved under:", save_root)
```

```python
N_TRIALS = 30
EPOCHS = 15
EARLY_STOPPING_PATIENCE = 30
MIN_DELTA = 0.0
RANDOM_SEED = 42

# =========================================================
# experiment settings
# =========================================================
rcm_var = "tmean"   # 'tmean' or 'prec'
gcm_name = "CanESM2"
gcm_rcm_dict = {"CanESM2": "CanRCM4"}
grid = "NAM-44i"
factor = 4

exp_folder_name = (
    f"/projects/sds-lab/Shuochen/downscaling/cordex_canesm2/"
    f"{rcm_var}.{gcm_name}.{gcm_rcm_dict[gcm_name]}.day.{grid}.mbcn-gridMET/"
)

# =========================================================
# temporal domain adaptation split
# source = past, target = future
# =========================================================
SOURCE_FRACTION = 0.7          # first 70% of time = source (past)
SOURCE_TRAIN_FRACTION = 0.8    # within source, first 80% = train, last 20% = val
TARGET_TRAIN_FRACTION = 0.8    # within target, first 80% = unsupervised target train, last 20% = target eval only

import os
import copy
import json
import math
import torch
import optuna
import numpy as np
from itertools import cycle
from torch import nn
from torch.utils.data import DataLoader, TensorDataset
from torch.optim.lr_scheduler import ReduceLROnPlateau
from torch.autograd import Function

from models import SRResNet_DANN

# =========================================================
# device
# =========================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

torch.manual_seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(RANDOM_SEED)

# =========================================================
# save paths
# =========================================================
model_name = "SRResNet_DANN"
save_root = os.path.join(exp_folder_name, "trained_models", model_name)
os.makedirs(save_root, exist_ok=True)

study_summary_path = os.path.join(save_root, "optuna_best_params.json")
all_trials_summary_path = os.path.join(save_root, "optuna_all_trials_summary.json")

print("All trial checkpoints will be saved to:", save_root)

# =========================================================
# load tensors
# =========================================================
X = torch.load(os.path.join(exp_folder_name, f"coarse_{factor}x.pth"))
y = torch.load(os.path.join(exp_folder_name, "high_res.pth"))
mask_hr = torch.load(os.path.join(exp_folder_name, "high_res_mask.pth"))

print("Loaded shapes:")
print("X       :", X.shape)
print("y       :", y.shape)
print("mask_hr :", mask_hr.shape)

if y.shape != mask_hr.shape:
    raise ValueError(
        f"Shape mismatch: y shape {y.shape} does not match mask shape {mask_hr.shape}"
    )

mask_hr = mask_hr.float()
mask_hr = (mask_hr > 0.5).float()

# =========================================================
# temporal split: source = past, target = future
# IMPORTANT: assumes sample dimension is time-ordered
# =========================================================
n_total = X.shape[0]
n_source_total = int(SOURCE_FRACTION * n_total)

if n_source_total <= 0 or n_source_total >= n_total:
    raise ValueError("SOURCE_FRACTION creates empty source or target split.")

X_source_all = X[:n_source_total]
y_source_all = y[:n_source_total]
mask_source_all = mask_hr[:n_source_total]

X_target_all = X[n_source_total:]
y_target_all = y[n_source_total:]           # evaluation only
mask_target_all = mask_hr[n_source_total:]  # evaluation only

print("\nTemporal split:")
print("X_source_all     :", X_source_all.shape)
print("y_source_all     :", y_source_all.shape)
print("mask_source_all  :", mask_source_all.shape)
print("X_target_all     :", X_target_all.shape)
print("y_target_all     :", y_target_all.shape)
print("mask_target_all  :", mask_target_all.shape)

# =========================================================
# source split: train / val
# target split: train / eval
# =========================================================
n_source_train = int(SOURCE_TRAIN_FRACTION * X_source_all.shape[0])
n_target_train = int(TARGET_TRAIN_FRACTION * X_target_all.shape[0])

if n_source_train <= 0 or n_source_train >= X_source_all.shape[0]:
    raise ValueError("SOURCE_TRAIN_FRACTION creates empty source train or val split.")
if n_target_train <= 0 or n_target_train >= X_target_all.shape[0]:
    raise ValueError("TARGET_TRAIN_FRACTION creates empty target train or eval split.")

X_source_train = X_source_all[:n_source_train]
y_source_train = y_source_all[:n_source_train]
mask_source_train = mask_source_all[:n_source_train]

X_source_val = X_source_all[n_source_train:]
y_source_val = y_source_all[n_source_train:]
mask_source_val = mask_source_all[n_source_train:]

X_target_train = X_target_all[:n_target_train]
X_target_eval = X_target_all[n_target_train:]
y_target_eval = y_target_all[n_target_train:]             # not used in training
mask_target_eval = mask_target_all[n_target_train:]       # not used in training

print("\nSource train/val shapes:")
print("X_source_train   :", X_source_train.shape)
print("y_source_train   :", y_source_train.shape)
print("mask_source_train:", mask_source_train.shape)
print("X_source_val     :", X_source_val.shape)
print("y_source_val     :", y_source_val.shape)
print("mask_source_val  :", mask_source_val.shape)

print("\nTarget train/eval shapes:")
print("X_target_train   :", X_target_train.shape)
print("X_target_eval    :", X_target_eval.shape)
print("y_target_eval    :", y_target_eval.shape)
print("mask_target_eval :", mask_target_eval.shape)

# =========================================================
# normalization stats from SOURCE TRAIN only
# =========================================================
X_mean = X_source_train.mean()
X_std = X_source_train.std()
X_max = X_source_train.max()

y_source_train_land = y_source_train[mask_source_train > 0.5]
if y_source_train_land.numel() == 0:
    raise ValueError("No land points found in source training mask.")

y_mean = y_source_train_land.mean()
y_std = y_source_train_land.std()
y_max = y_source_train_land.max()

if X_std.item() == 0:
    X_std = torch.tensor(1.0, dtype=X_source_train.dtype)
if y_std.item() == 0:
    y_std = torch.tensor(1.0, dtype=y_source_train.dtype)

# normalize source with source stats
X_source_train_n = (X_source_train - X_mean) / X_std
X_source_val_n   = (X_source_val   - X_mean) / X_std
y_source_train_n = (y_source_train - y_mean) / y_std
y_source_val_n   = (y_source_val   - y_mean) / y_std

# normalize target with SOURCE stats
X_target_train_n = (X_target_train - X_mean) / X_std
X_target_eval_n  = (X_target_eval  - X_mean) / X_std

# =========================================================
# losses
# =========================================================
def masked_mse(pred, target, mask, eps=1e-8):
    """
    pred, target, mask: [B, C, H, W]
    mask = 1 over land, 0 over ocean
    Returns mean squared error over land pixels only.
    """
    se = (pred - target) ** 2
    se = se * mask
    return se.sum() / (mask.sum() + eps)

sr_loss_fn = masked_mse
val_loss_fn = nn.MSELoss()
domain_loss_fn = nn.CrossEntropyLoss()

# =========================================================
# datasets
# =========================================================
source_training_set = TensorDataset(
    X_source_train_n,
    y_source_train_n,
    X_source_train,
    y_source_train,
    mask_source_train
)

source_validation_set = TensorDataset(
    X_source_val_n,
    y_source_val_n,
    X_source_val,
    y_source_val,
    mask_source_val
)

target_training_set = TensorDataset(X_target_train_n)

target_eval_set = TensorDataset(
    X_target_eval_n,
    y_target_eval,
    mask_target_eval
)

# =========================================================
# helper for DANN lambda schedule
# =========================================================
def dann_lambda_schedule(progress):
    """
    progress in [0, 1]
    standard DANN schedule
    """
    return 2.0 / (1.0 + math.exp(-10.0 * progress)) - 1.0

# =========================================================
# objective function
# =========================================================
def objective(trial):
    # -----------------------------------------------------
    # hyperparameters to search
    # -----------------------------------------------------
    batch_size = trial.suggest_categorical("batch_size", [256, 512, 1024])
    learning_rate = trial.suggest_float("learning_rate", 1e-4, 3e-4, log=True)
    num_resblk = trial.suggest_int("num_resblk", 4, 16, step=2)
    num_features = trial.suggest_categorical("num_features", [32, 64, 96, 128])
    weight_decay = trial.suggest_float("weight_decay", 1e-6, 1e-3, log=True)
    domain_hidden_dim = trial.suggest_categorical("domain_hidden_dim", [64, 128, 256])
    alpha_domain = trial.suggest_float("alpha_domain", 0.01, 1.0, log=True)

    # -----------------------------------------------------
    # paths for this trial
    # -----------------------------------------------------
    trial_ckpt_path = os.path.join(save_root, f"trial_{trial.number:03d}.pth")

    # -----------------------------------------------------
    # model
    # -----------------------------------------------------
    model = SRResNet_DANN(
        num_resblk=num_resblk,
        num_features=num_features,
        input_channels=1,
        output_channels=1,
        scale=factor,
        num_domains=2,
        domain_hidden_dim=domain_hidden_dim,
        lambda_grl=1.0,
    ).to(device)

    # -----------------------------------------------------
    # dataloaders
    # -----------------------------------------------------
    source_train_dataloader = DataLoader(
        source_training_set,
        batch_size=batch_size,
        shuffle=True,
        drop_last=True,
    )

    target_train_dataloader = DataLoader(
        target_training_set,
        batch_size=batch_size,
        shuffle=True,
        drop_last=True,
    )

    source_val_dataloader = DataLoader(
        source_validation_set,
        batch_size=batch_size,
        shuffle=False,
    )

    target_eval_dataloader = DataLoader(
        target_eval_set,
        batch_size=batch_size,
        shuffle=False,
    )

    if len(source_train_dataloader) == 0:
        raise ValueError("Source train dataloader is empty. Reduce batch size or check data size.")
    if len(target_train_dataloader) == 0:
        raise ValueError("Target train dataloader is empty. Reduce batch size or check data size.")

    # -----------------------------------------------------
    # training setup
    # -----------------------------------------------------
    optimizer = torch.optim.AdamW(
        model.parameters(),
        lr=learning_rate,
        weight_decay=weight_decay,
    )
    scheduler = ReduceLROnPlateau(optimizer, patience=15, factor=0.5)

    train_total_loss_list = []
    train_sr_loss_list = []
    train_domain_loss_list = []
    source_val_loss_list = []
    target_eval_loss_list = []

    best_val_loss = float("inf")
    best_epoch = -1
    epochs_no_improve = 0

    y_mean_dev = y_mean.to(device)
    y_std_dev = y_std.to(device)

    total_training_steps = EPOCHS * max(len(source_train_dataloader), len(target_train_dataloader))
    global_step = 0

    # -----------------------------------------------------
    # train loop
    # -----------------------------------------------------
    for epoch in range(EPOCHS):
        model.train()

        epoch_total_loss = 0.0
        epoch_sr_loss = 0.0
        epoch_domain_loss = 0.0
        num_batches = 0

        target_iter = cycle(target_train_dataloader)

        for Xs_n, ys_n, Xs_raw, ys_raw, mask_s_raw in source_train_dataloader:
            Xt_n = next(target_iter)[0]

            Xs_n = Xs_n.to(device)
            ys_n = ys_n.to(device)
            mask_s_raw = mask_s_raw.to(device)

            Xt_n = Xt_n.to(device)

            progress = global_step / max(total_training_steps - 1, 1)
            lambda_grl = dann_lambda_schedule(progress)

            # source forward
            ys_pred_n, domain_logits_s = model(Xs_n, lambda_grl=lambda_grl)

            # target forward
            _, domain_logits_t = model(Xt_n, lambda_grl=lambda_grl)

            # SR loss on source only
            ys_pred_n = ys_pred_n * mask_s_raw
            loss_sr = sr_loss_fn(ys_pred_n, ys_n, mask_s_raw)

            # domain loss
            domain_labels_s = torch.zeros(
                Xs_n.size(0), dtype=torch.long, device=device
            )  # 0 = past/source
            domain_labels_t = torch.ones(
                Xt_n.size(0), dtype=torch.long, device=device
            )  # 1 = future/target

            loss_domain_s = domain_loss_fn(domain_logits_s, domain_labels_s)
            loss_domain_t = domain_loss_fn(domain_logits_t, domain_labels_t)
            loss_domain = loss_domain_s + loss_domain_t

            # total loss
            loss = loss_sr + alpha_domain * loss_domain

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            epoch_total_loss += loss.item()
            epoch_sr_loss += loss_sr.item()
            epoch_domain_loss += loss_domain.item()
            num_batches += 1
            global_step += 1

        train_total_loss = epoch_total_loss / max(num_batches, 1)
        train_sr_loss = epoch_sr_loss / max(num_batches, 1)
        train_domain_loss = epoch_domain_loss / max(num_batches, 1)

        train_total_loss_list.append(train_total_loss)
        train_sr_loss_list.append(train_sr_loss)
        train_domain_loss_list.append(train_domain_loss)

        # -----------------------------------------------------
        # validation on source only
        # full-image MSE in physical space
        # -----------------------------------------------------
        model.eval()
        source_val_loss = 0.0

        with torch.no_grad():
            for Xs_n, ys_n, Xs_raw, ys_raw, mask_s_raw in source_val_dataloader:
                Xs_n = Xs_n.to(device)
                ys_raw = ys_raw.to(device)
                mask_s_raw = mask_s_raw.to(device)

                ys_pred_n, _ = model(Xs_n, lambda_grl=0.0)
                ys_pred = ys_pred_n * y_std_dev + y_mean_dev

                # exact zero over ocean
                ys_pred = ys_pred * mask_s_raw

                # full-image MSE
                source_val_loss += val_loss_fn(ys_pred, ys_raw).item()

        source_val_loss /= len(source_val_dataloader)
        source_val_loss_list.append(source_val_loss)

        # -----------------------------------------------------
        # target evaluation (future climate)
        # NOT used for training or early stopping
        # -----------------------------------------------------
        target_eval_loss = 0.0

        with torch.no_grad():
            for Xt_n, yt_raw, mask_t_raw in target_eval_dataloader:
                Xt_n = Xt_n.to(device)
                yt_raw = yt_raw.to(device)
                mask_t_raw = mask_t_raw.to(device)

                yt_pred_n, _ = model(Xt_n, lambda_grl=0.0)
                yt_pred = yt_pred_n * y_std_dev + y_mean_dev

                # exact zero over ocean
                yt_pred = yt_pred * mask_t_raw

                # future full-image MSE
                target_eval_loss += val_loss_fn(yt_pred, yt_raw).item()

        target_eval_loss /= len(target_eval_dataloader)
        target_eval_loss_list.append(target_eval_loss)

        scheduler.step(source_val_loss)

        # -----------------------------------------------------
        # report to Optuna using SOURCE validation only
        # -----------------------------------------------------
        trial.report(source_val_loss, epoch)
        trial.set_user_attr("best_epoch", best_epoch)
        trial.set_user_attr("best_val_loss", best_val_loss)

        if trial.should_prune():
            trial.set_user_attr("status", "pruned")
            raise optuna.TrialPruned()

        # -----------------------------------------------------
        # save best model using SOURCE validation only
        # -----------------------------------------------------
        if source_val_loss < best_val_loss - MIN_DELTA:
            best_val_loss = source_val_loss
            best_epoch = epoch
            epochs_no_improve = 0

            trial.set_user_attr("best_epoch", best_epoch)
            trial.set_user_attr("best_val_loss", best_val_loss)
            trial.set_user_attr("checkpoint_path", trial_ckpt_path)
            trial.set_user_attr("best_target_eval_loss", target_eval_loss)

            torch.save(
                {
                    "trial_number": trial.number,
                    "epoch": best_epoch,
                    "model_name": model.__class__.__name__,
                    "model_hparams": {
                        "num_resblk": num_resblk,
                        "num_features": num_features,
                        "input_channels": 1,
                        "output_channels": 1,
                        "scale": factor,
                        "num_domains": 2,
                        "domain_hidden_dim": domain_hidden_dim,
                    },
                    "training_hparams": {
                        "batch_size": batch_size,
                        "learning_rate": learning_rate,
                        "weight_decay": weight_decay,
                        "alpha_domain": alpha_domain,
                        "epochs": EPOCHS,
                        "early_stopping_patience": EARLY_STOPPING_PATIENCE,
                        "min_delta": MIN_DELTA,
                        "loss": "source_masked_mse + alpha_domain * domain_ce",
                        "validation_loss": "source_full_image_mse",
                        "target_eval_loss": "future_full_image_mse_not_used_for_early_stopping",
                        "ocean_constraint": "prediction multiplied by land mask in physical space",
                        "domain_definition": "source=past_climate, target=future_climate",
                        "source_fraction": SOURCE_FRACTION,
                        "source_train_fraction": SOURCE_TRAIN_FRACTION,
                        "target_train_fraction": TARGET_TRAIN_FRACTION,
                    },
                    "model_state": copy.deepcopy(model.state_dict()),
                    "optimizer_state": optimizer.state_dict(),
                    "scheduler_state": scheduler.state_dict(),
                    "best_val_loss": best_val_loss,
                    "best_epoch": best_epoch,
                    "source_val_loss": source_val_loss,
                    "target_eval_loss": target_eval_loss,
                    "train_total_loss": train_total_loss,
                    "train_sr_loss": train_sr_loss,
                    "train_domain_loss": train_domain_loss,
                    "train_total_loss_list": train_total_loss_list,
                    "train_sr_loss_list": train_sr_loss_list,
                    "train_domain_loss_list": train_domain_loss_list,
                    "source_val_loss_list": source_val_loss_list,
                    "target_eval_loss_list": target_eval_loss_list,
                    "X_mean": X_mean,
                    "X_std": X_std,
                    "X_max": X_max,
                    "y_mean": y_mean,
                    "y_std": y_std,
                    "y_max": y_max,
                },
                trial_ckpt_path,
            )
        else:
            epochs_no_improve += 1

        print(
            f"Trial {trial.number:03d} | "
            f"Epoch {epoch:03d} | "
            f"Train Total: {train_total_loss:.6f} | "
            f"Train SR: {train_sr_loss:.6f} | "
            f"Train Domain: {train_domain_loss:.6f} | "
            f"Source Val: {source_val_loss:.6f} | "
            f"Target Eval: {target_eval_loss:.6f} | "
            f"Best Source Val: {best_val_loss:.6f} | "
            f"Best Epoch: {best_epoch:03d} | "
            f"lambda_grl: {lambda_grl:.4f} | "
            f"No improve: {epochs_no_improve}/{EARLY_STOPPING_PATIENCE} | "
            f"lr: {optimizer.param_groups[0]['lr']:.6e}"
        )

        if epochs_no_improve >= EARLY_STOPPING_PATIENCE:
            print(
                f"Early stopping triggered for trial {trial.number:03d} "
                f"at epoch {epoch:03d} | best epoch: {best_epoch:03d}"
            )
            break

    trial.set_user_attr("status", "completed")
    trial.set_user_attr("best_epoch", best_epoch)
    trial.set_user_attr("best_val_loss", best_val_loss)
    trial.set_user_attr("best_target_eval_loss", target_eval_loss_list[best_epoch] if len(target_eval_loss_list) > 0 and best_epoch >= 0 else None)
    trial.set_user_attr("checkpoint_path", trial_ckpt_path)

    return best_val_loss


# =========================================================
# create and run study
# =========================================================
study = optuna.create_study(
    direction="minimize",
    pruner=optuna.pruners.MedianPruner(n_startup_trials=5, n_warmup_steps=5),
)

study.optimize(objective, n_trials=N_TRIALS)

# =========================================================
# save best params summary
# =========================================================
best_trial = study.best_trial
best_summary = {
    "best_trial_number": best_trial.number,
    "best_value": best_trial.value,
    "best_params": best_trial.params,
    "best_epoch": best_trial.user_attrs.get("best_epoch", None),
    "best_target_eval_loss": best_trial.user_attrs.get("best_target_eval_loss", None),
    "best_checkpoint_path": best_trial.user_attrs.get(
        "checkpoint_path",
        os.path.join(save_root, f"trial_{best_trial.number:03d}.pth")
    ),
}

with open(study_summary_path, "w") as f:
    json.dump(best_summary, f, indent=4)

# =========================================================
# save all trial summaries
# =========================================================
all_trials_summary = []
for t in study.trials:
    trial_info = {
        "trial_number": t.number,
        "state": str(t.state),
        "value": t.value if t.value is not None else None,
        "params": t.params,
        "best_epoch": t.user_attrs.get("best_epoch", None),
        "best_val_loss": t.user_attrs.get("best_val_loss", None),
        "best_target_eval_loss": t.user_attrs.get("best_target_eval_loss", None),
        "checkpoint_path": t.user_attrs.get(
            "checkpoint_path",
            os.path.join(save_root, f"trial_{t.number:03d}.pth")
        ),
    }
    all_trials_summary.append(trial_info)

with open(all_trials_summary_path, "w") as f:
    json.dump(all_trials_summary, f, indent=4)

# =========================================================
# final logs
# =========================================================
print("\nOptuna study finished.")
print(f"Best trial: {best_trial.number}")
print(f"Best value (source val): {best_trial.value:.6f}")
print(f"Best epoch: {best_trial.user_attrs.get('best_epoch', None)}")
print(f"Best target eval loss: {best_trial.user_attrs.get('best_target_eval_loss', None)}")
print("Best params:", best_trial.params)
print(
    "Best checkpoint:",
    best_trial.user_attrs.get(
        "checkpoint_path",
        os.path.join(save_root, f"trial_{best_trial.number:03d}.pth")
    )
)
print("Best summary saved to:", study_summary_path)
print("All trial summaries saved to:", all_trials_summary_path)
print("All per-trial checkpoints saved under:", save_root)
```
